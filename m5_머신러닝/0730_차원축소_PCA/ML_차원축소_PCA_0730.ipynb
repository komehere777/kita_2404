## 다중공선성(Multicollinearity)
회귀 분석에서 사용되는 독립 변수들 간에 높은 상관관계가 있을 때 발생하는 현상. 이는 한 독립 변수가 다른 하나 또는 그 이상의 독립 변수와 선형적으로 매우 밀접하게 관련되어 있음을 의미.(상관관계가 매우 높다는 것)

다중공선성이 존재하면, 회귀 분석의 결과를 해석하는 데 어려움이 생기며, 모델의 예측 성능에 부정적인 영향을 줄 수 있다.

다중공선성의 주요 문제점:
- 계수 추정의 불안정성: 독립 변수들 간의 높은 상관관계는 회귀 계수(회귀 모델에서 독립 변수의 `가중치`)의 추정치가 불안정하게 되어- 모델의 가치가 떨어짐, 작은 데이터 변경에 의해 추정치가 크게 변동할 수 있다.

- 해석의 어려움: 다중공선성이 있는 경우, 각 독립 변수의 영향을 개별적으로 분리하여 해석하기 어렵다. 즉, 어떤 변수가 종속 변수에 미치는 영향을 정확히 파악하기 어렵다.

- 과적합(Overfitting)의 위험성: 모델이 학습 데이터에는 잘 맞지만, 새로운 데이터에 대해서는 예측 성능이 떨어지는 과적합 현상이 발생할 수 있다.

- 다중공선성의 진단:
  - 분산팽창인자(Variance Inflation Factor, VIF): VIF는 독립 변수가 다른 독립 변수들에 의해 얼마나 설명되는지를 측정하는 지표. 일반적으로 VIF 값이 10 이상이면 해당 변수는 다른 변수와 높은 다중공선성을 가진다고 판단.
  - 상관 행렬(Correlation Matrix): 독립 변수들 간의 상관 관계를 확인하는 방법으로, 높은 상관계수(가이드라인: 절대값 기준 0.8 이상)는 다중공선성의 가능성을 시사. - 사안별로 파악해야함.

- 다중공선성의 해결 방법:
  - 변수 제거: 상관관계가 높은 변수들 중 하나 또는 여러 개를 모델에서 제거.
  - 주성분 분석(PCA): 여러 독립 변수를 적은 수의 비상관 변수로 변환하는 기법으로, 변환된 변수를 사용하여 모델링.
  - 릿지(Ridge) 또는 라쏘(Lasso) 회귀: 정규화 기법을 사용하여 회귀 계수의 크기를 줄이고, 과적합을 방지.

다중공선성은 모델의 신뢰성과 정확성에 영향을 미치므로, 회귀 분석을 수행할 때 주의 깊게 고려되어야 한다.

## 차원 축소
* 차원 축소의 목적: 다중공선성이 발생하면 차원 축소를 통해 불필요한 피처를 제거해야 함
* 매우 많은 피처로 구성된 다차원 데이터 세트의 차원을 축소해 새로운 차원의 데이터 세트를 생성하는 것
* 차원(컬럼수)이 증가할 수록 데이터 포인트 간의 거리가 기하급수적으로 멀어지게 되고 희소한 구조를 가지며 예측 신뢰도가 떨어짐
* 선형 모델에서는 입력 변수 간의 상관관계가 높을 경우 다중 공선성 문제로 모델의 예측 성능 저하 - 회귀계수가 부정확
* 다차원의 피처를 차원 축소해 피처 수를 줄이면 더 직관적으로 데이터를 해석할 수 있음
* 차원 축소는 피처 선택과 피처 추출로 나눌 수 있음
* 피처 선택은 특정 피처에 종속성이 강한 불필요한 피처는 제거하고 데이터의 특징을 잘 나타내는 주요 피처만 선택(변수 줄이기- 물리적 축소)
* 피처 추출은 기존 피처를 저차원의 중요 피처로 압축해서 추출. 기존의 피처가 압축된 것이므로 기존 피처와는 다른 값이 됨(변수 변환- 화학적 축소)
* 차원 축소를 통해 좀 더 데이터를 잘 설명할 수 있는 잠재적인 요소(축소과정의 매개체에 의해서 축소 , 그 매개체가 잠재적인 요소) 를 추출하며 PCA, SVD, NMF은 대표적인 차원 축소 알고리즘

## PCA
* 입력 데이터의 공분산 행렬이 `고유벡터`와 `고유값`으로 분해될 수 있으며 이렇게 분해된 고유벡터을 이용해 입력 데이터를 선형 변환하는 방식이 PCA
* 입력 데이터 세트의 공분산 행렬을 생성
* 공분산 행렬의 고유벡터와 고유값을 계산
* 공분산 C = 고유벡터 직교행렬 * 고유값 정방행렬 * 고유벡터 직교 행렬의 전치행렬
* 고유값이 가장 큰 순으로 K개(PCA 변환 차수)만큼 고유벡터를 추출
* 고유값이 가장 큰 순으로 추출되며 고유벡터를 이용해 새롭게 입력 데이터를 변환
* 입력 데이터의 공분산 행렬이 고유벡터와 고유값으로 분해될 수 있으며 이렇게 분해된 고유벡터를 이용해 입력데이터를 선형 변환하는 방식이 PCA
  1. 입력 데이터 세트의 공분산 행렬을 생성
  2. 공분산 행렬의 고유벡터와 고유값을 계산
  3. 고유값이 가장 큰 순으로 K개(PCA 변환 차수만큼) 고유벡터로 추출
  4. 고유값이 가장 큰 순으로 추출된 고유벡터을 이용해 새롭게 입력 데이터를 변환

## 선형대수식
* 크기와 방향을 가지고 있는 것을 vector, 반면 크기만 있는 것(질량, 온도, 길이)을 scalar이라고 한다
* 고유벡터 : 행렬에 의해 변환될 때 방향(분산의 변동성)이 변하지 않고 오직 크기만 변하는 벡터 Ax = ax(A 행렬, x 고유벡터, a 스칼라(고유값))
* 고유값(Eigenvalue) : 행렬이 벡터에 작용할 때, 그 벡터의 방향은 유지하면서 스케일(크기)만 변화시키는 스케일링 요소
* 고유값과 고유벡터는 행렬이 벡터에 작용하는 방식을 설명하는데 사용되며, 행렬이 벡터를 어떻게 스케일링하고 방향을 유지하는지를 나타낸다.
* 선형 변환 : 특정 벡터에 행렬  A를 곱해 새로운 벡터로 변환
* 공분산 행렬 : 두변수간의 변동을 의미. 공분산 Cov(X,Y) > 0 은 X가 증가할 때 Y도 증가
* 대칭행렬 : 고유벡터를 항상 직교행렬로 고유값을 정방 행렬로 대각화할 수 있음
* 직교행렬 : 두 벡터가 직교한다는 것은 그들이 서로 수직을 이룬다는 의미.대각원소 이외의 모든 원소가 0일때. 직교행렬의 역행렬은 직교행렬 자신의 전치행렬과 같음. 직교행렬의 모든 행 벡터와 열 벡터는 단위 벡터(크기가 1인 벡터, 방향을 나타내지만 크기를 고려하지 않는 벡터)이며 서로 직교

## LDA(Linear Discriminant Analysis)
* PCA와 매우 유사한 방식이며 PCA가 입력 데이터 `변동성의 가장 큰 축`을 찾는데(분산이 큰) 반해 LDA는 입력 데이터의 결정 값 `클래스를 최대한으로 분리`할 수 있는 축을 찾는 방식으로 차원을 축소

## SVD, NMF 
* SVD(Singular Value Decomposition) : PCA와 유사한 행렬 분해 기법. 정방행렬뿐만 아니라 행과 열의 크기가 다른 행렬에도 적용 가능. SVD는 특이값 분해로 불림.
* 매우 많은 피처 데이터를 가진 고차원 행렬을 두 개의 저차원 행렬로 분리하는 행렬 분해 기법.
* 행렬 분해를 수행하면서 원본 행렬에서 잠재된 요소를 추출하기 때문에 토픽 모델이나 추천 시스템에서 활발하게 사용

```python
import matplotlib.pyplot as plt

image_path = "data\eigenvector.png"
# 이미지 불러오기
img = plt.imread(image_path)
plt.figure(figsize=(15, 12))

# 이미지 표시
plt.imshow(img)
plt.axis("off")  # 축 제거
plt.show()
```

```python
import numpy as np

A = np.array([[4, 1],
              [2, 3]])

eigenvalues, eigenvectors = np.linalg.eig(A)

print("eigenvalues:", eigenvalues)
print("eigenvectors:\n", eigenvectors)
```

eigenvalues: 5 가 크므로 주성분

```python
#첫 번째 고유값과 고유벡터
lambda_1 = eigenvalues[0]  # 스칼라 값
v_1 = eigenvectors[:, 0]

# 검증
Av_1 = np.dot(A, v_1)  #벡터 A와 v_1의 행렬곱이므로 np.dot() 사용
lambda_1_v_1 = lambda_1 * v_1 # lambda_1은 스칼라이므로 곱셈 순서 바뀌어도 상관 없음, 스칼라 곱셈은 그냥 곱셈

print("A * v_1:", Av_1)
print("lambda_1 * v_1:", lambda_1_v_1)
print("Are they same?", np.allclose(Av_1, lambda_1_v_1))

lambda_2 = eigenvalues[1]
v_2 = eigenvectors[:, 1]

Av_2 = np.dot(A, v_2)
lambda_2_v_2 = lambda_2 * v_2

print("A * v_2:", Av_2)
print("lambda_2 * v_2:", lambda_2_v_2)
print("Are they same?", np.allclose(Av_2, lambda_2_v_2))
```

```python
# PCA 분석
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt

iris = load_iris()
# 넘파이 데이터 세트를 판다스 DataFrame으로 변환
columns = ["sepal_length", "sepal_width", "petal_length", "petal_width"]
df_iris = pd.DataFrame(iris.data, columns=columns)
df_iris["target"] = iris.target
df_iris.head()
```

matplotlib에서는 다음과 같은 다양한 마커 옵션을 제공:

- 'o': 원(circle)
- '^': 위쪽을 향한 삼각형(triangle_up)
- 's': 사각형(square)
- '+': 플러스(plus sign)
- '.': 점(point)
- '*': 별표(별)
- 'x': 엑스(x)
- '|': 세로 막대(vertical line)
-'_': 가로 막대(horizontal line)

```python
# sepal_length와 sepal_width 두 개의 속성만 사용, 품종 데이터 분포 시각화
# setosa는 세모, versicolor는 네모, virginica는 동그라미로 표현

markers = ["^", "s", "o"]

# setosa의 target 값은 0, versicolor는 1, virginica는 2. 각 target 별로 다른 모양으로 산점도로 표시
# 산점도를 이용해 시각화
for i, marker in enumerate(markers):
    x_axis_data = df_iris[df_iris["target"] == i]["sepal_length"]
    y_axis_data = df_iris[df_iris["target"] == i]["sepal_width"]
    plt.scatter(x_axis_data, y_axis_data, marker=marker, label=iris.target_names[i])
    
plt.legend()
plt.xlabel("sepal length")
plt.ylabel("sepal width")
```

세토사는 잘 구분되나 , 버지니카와 베르시칼라는 구분이 어려움

```python
df_iris1 = df_iris.drop("target", axis=1)

# standard scaling 적용, 평균0, 분산이 1인 표준 정규 분포로 모든 속성값 변환
from sklearn.preprocessing import StandardScaler
iris_scaled = StandardScaler().fit_transform(df_iris1)
```

```python
# PCA 로 4개 속성을 2개로 압축
from sklearn.decomposition import PCA

# 생성 파라미터로 n_components를 입력 받음, PCA로 변환할 차원의 수를 의미
pca = PCA(n_components=2)

# fit( )과 transform( )을 호출해 PCA 변환 데이터 반환
iris_pca = pca.fit_transform(iris_scaled)
print(iris_pca.shape)
```

```python
type(iris_pca)  #배열 타입
```

```python
iris_pca[:5]  #주성분2ㄱ개로 변환된 데이터
```

```python
# 4개가 2개로 줄어듬 확인, 화학적 축소

pca_columns = ["pca_component_1", "pca_component_2"]
df_iris_pca = pd.DataFrame(iris_pca, columns=pca_columns)
df_iris_pca["target"] = iris.target
df_iris_pca.head()
```

```python
markers = ["^", "s", "o"]

# setosa의 target 값은 0, versicolor는 1, virginica는 2. 각 target 별로 다른 모양으로 산점도로 표시
# 산점도를 이용해 시각화
for i, marker in enumerate(markers):
    x_axis_data = df_iris_pca[df_iris_pca["target"] == i]["pca_component_1"]
    y_axis_data = df_iris_pca[df_iris_pca["target"] == i]["pca_component_2"]
    plt.scatter(x_axis_data, y_axis_data, marker=marker, label=iris.target_names[i])

plt.legend()
plt.xlabel("pca_component_1")
plt.ylabel("pca_component_2")
```

대표적인 두개의 컬럼으로 한 것보다 주성분 분석를 사용한 것이 더 잘 구분됨

```python
# c1, c2가 전체 변동성을 얼마나 설명하는지 확인
# explained_variance_ratio_ : 전체 변동성에서 개별 PCA 컴포넌트별로 차지하는 변동성 비율 제공
# c1 : 0.7296, c2 : 0.2285, 합 0.9581
# 2개의 주성분으로 95.81%의 변동성을 설명할 수 있음
pca.explained_variance_ratio_  #각 주성분이 설명하는 분산의 비율
```
