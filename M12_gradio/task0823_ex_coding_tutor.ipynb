````python
import gradio as gr
from openai import OpenAI

client = OpenAI()

system_msg = """
• 코딩튜터: 고등학생 정보과목 알고리즘/프로그래밍 실습 도우미
• 인사: "반가와요~"로 시작
• 학습 시작: 학생 선택 섹션 요약 제공
• 학습 방식:
  - 개념 설명: 명확하고 친절하게
  - 연습문제: 개념 이해와 활용 위해 제공
• 피드백: 
  - "잘했어요!", "조금 더 생각해봐요~" 등 긍정적
  - "힘내세요!", "할 수 있어요!" 등 동기부여 메시지 추가
• 대화 길이: 
  - 일반 텍스트 최대 100자
  - 코드 최대 300자
• 특징: 
  - 자기주도 학습 지원
  - 개인화된 학습 제공
  - 한국어로 진행
  - 주기적으로 격려와 응원 메시지 전달
  """

def chat_with_tutor(user_input, chat_history):
    messages = [{"role": "system", "content": system_msg}] + chat_history

    if not chat_history:
        messages.append(
            {
                "role": "assistant",
                "content": "안녕하세요. 파이썬 기초에 대하여 학습을 지도해줘요",
            }
        )

    messages.append({"role": "user", "content": user_input})

    response_ = client.chat.completions.create(
        model="gpt-4o-mini-2024-07-18",
        messages=messages,
        temperature=0.5,
        max_tokens=300,
        top_p=1.0,
    )
    response_msg = response_.choices[0].message.content
    chat_history.append({"role": "user", "content": user_input})
    chat_history.append({"role": "assistant", "content": response_msg})
    return response_msg, chat_history, update_chat_history(chat_history)


def start_chat():
    initial_message = "반가와요~ 파이썬 기초에 대해서 학습을 시작해볼까요? 먼저 파이썬의 기본 문법과 개념들을 간단히 요약해드릴께요"
    initial_history = [{"role": "assistant", "content": initial_message}]
    return initial_message, initial_history, update_chat_history(initial_history)

def update_chat_history(history):
    if history:
        return "\n".join([f"{msg['role']}: {msg['content']}" for msg in history])
    return ""


with gr.Blocks() as demo:
    gr.Markdown("# 파이썬 튜터 봇")

    chat_history = gr.State([])

    with gr.Row():
        start_btn = gr.Button("Start Chat")

    with gr.Row():
        with gr.Column(scale=1, min_width=300):
            gr.Markdown("History")
            chat_history_display = gr.Textbox(
                label="Chat History", type="text", lines=12
            )
        with gr.Column(scale=2, min_width=300):
            gr.Markdown("Chat")
            tutor_response = gr.Textbox(lines=5, label="Tutor")
            user_input = gr.Textbox(lines=5, label="User")

    with gr.Row():
        submit_btn = gr.Button("Submit")

    start_btn.click(
        start_chat, outputs=[tutor_response, chat_history, chat_history_display]
    )
    submit_btn.click(
        chat_with_tutor,
        inputs=[user_input, chat_history],
        outputs=[tutor_response, chat_history, chat_history_display],
    )

demo.launch(debug=True)
````

````python
import gradio as gr
from openai import OpenAI

# OpenAI API 설정
client = OpenAI()

# 시스템 메시지 설정
system_msg = """코딩튜터는 고등학생 정보과목에서 알고리즘과 프로그래밍을 학생들이 실습할 수 있게 도와주는 튜터봇입니다.
              - 튜터는 "반가와요~ "라고 인사말로 학습을 시작합니다.
              - 학생이 선택한 섹션으로 학습을 시작하고 먼저 간단하게 섹션 학습내용 요약을 제공합니다.
              - 각 섹션별 학습 항목에 대해서 하나씩 개념을 명확하고 친절하게 설명하고 연습문제를 추가로 제공해서 각 개념을 이해하여 활용할 수 있도록 연습 기회를 제공합니다.
              - 학생의 답변에 대해서 "잘했어요!","조금 더 생각해봐요~" 와 같은 긍정적인 피드백을 제공합니다.
              - 대화는 간결하게 유지하며, 최대 100자 이내로 진행합니다.단 코드는 300자 이내로 작성합니다.
              - 튜터는 학생들이 자기주도 학습을 할 수 있도록 개인화 된 학습을 제공하고 모든 질의 응답을 한국어로 진행합니다."""


# 채팅 함수
def chat_with_tutor(user_input, chat_history):
    messages = [{"role": "system", "content": system_msg}] + chat_history

    if not chat_history:  # 첫 대화일 경우 인사말과 함께 시작
        messages.append({"role": "user", "content": "파이썬 학습을 시작해줘요"})

    messages.append({"role": "user", "content": user_input})

    response = client.chat.completions.create(
        model="gpt-4o-mini-2024-07-18",
        messages=messages,
        temperature=0.5,
        max_tokens=300,
        top_p=1,
    )

    response_msg = response.choices[0].message.content
    chat_history.append({"role": "assistant", "content": response_msg}) # 대화 이력에 추가

    return response_msg, chat_history


# Gradio UI 구성
def start_chat():
    return (
        "반가워요~ 파이썬 기초를 학습해보도록 할게요. 먼저 파이썬의 기본 문법과 개념들을 간단히 요약해드릴게요.",
        [],
    )


with gr.Blocks() as demo:
    gr.Markdown("# 파이썬 튜터 봇")

    chat_history = gr.State([])

    with gr.Row():
        start_button = gr.Button("학습 시작")

    tutor_response = gr.Textbox(label="튜터")
    user_input = gr.Textbox(label="학생")

    with gr.Row():
        submit_button = gr.Button("제출")

    # 그라디오는 함수 실행 후 반환된는 값을 지정된 ui요소에 자동으로 할당
    start_button.click(start_chat, inputs=[], outputs=[tutor_response, chat_history])
    submit_button.click(
        chat_with_tutor,
        inputs=[user_input, chat_history],
        outputs=[tutor_response, chat_history],  # 그라디오 UI에 출력
    )

# 실행
demo.launch(debug=True)
````

## web ui 에서 input 함수 처리

````python
import re
import gradio as gr

def execute_code(code, user_input_value):
    try:
        if user_input_value:
            # 사용자 입력값을 코드에 삽입
            code = re.sub(r"input\((.*?)\)", f'"{user_input_value}"', code)
        print(f"대체된 코드: {code}")
        
        exec_globals = {}
        exec_output = []
        
        def custom_print(*args):
            exec_output.append(" ".join(map(str, args)))
            
        exec_globals["print"] = custom_print
        
        
        for k, y in exec_globals.items():
            print(f"전달된 전역 변수: {k}, {y}")
        
        exec(code, exec_globals)
        
        for j in exec_output:
            print(f'출력: {j}')
        
        exec_output_str = "\n".join(exec_output) if exec_output else "코드 실행 성공!"
        return exec_output_str
    
    except Exception as e:
        return f"코드 실행 오류: {str(e)}"
    
def check_for_input_function(code):
    return bool(re.search(r"input\((.*?)\)", code))

with gr.Blocks() as demo:
    code_input = gr.Textbox(label="파이썬 코드 입력", placeholder="input()을 포함한 코드를 입력하세요", lines=3)
    user_input_value = gr.Textbox(label="input() 값 입력후 엔터!!", placeholder="input() 함수에 전달할 값 입력", visible=False)
    code_output = gr.Textbox(label="코드 실행 결과", placeholder="코드 실행 결과가 여기에 표시됩니다", lines=3)

    code_run_button = gr.Button("코드 실행")
    
    def handle_code_execution(code):
        if check_for_input_function(code):
            return gr.update(visible=True), ""  # 사용자 입력값 입력 필드 표시, 실행결과 없음
        else:
            output = execute_code(code, None)  # 사용자 입력값 없이 코드 실행
            return gr.update(visible=False), output  # 사용자 입력값 입력 필드 숨김, 실행결과 반환
        
    def excute_with_input(code, user_input_value):
        output = execute_code(code, user_input_value)
        return output
    
    code_run_button.click(
        handle_code_execution,
        inputs=[code_input],
        outputs=[user_input_value, code_output]
    )
    
    user_input_value.submit(
        excute_with_input,
        inputs=[code_input, user_input_value],
        outputs=[code_output]
    )
    
demo.launch(debug=True)
                        
````
