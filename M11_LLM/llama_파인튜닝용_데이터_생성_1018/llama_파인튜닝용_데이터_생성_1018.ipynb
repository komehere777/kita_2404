```python
import pandas as pd
import json
import jsonlines
import numpy as np
import os
from datasets import Dataset

```

```python
datasetName = "train_1.csv"
jsonFileName = "train_1.jsonl"
```

```python
def csv_to_json(csv_file_path, json_file_path):
    df = pd.read_csv(csv_file_path, encoding='utf-8')

    with open(json_file_path, 'w', encoding='utf-8') as json_file:
        for index, row in df.iterrows():
            data = {'inputs': row['질문'], 'responses': row['답변']}
            json.dump(data, json_file, ensure_ascii=False)
            json_file.write('\n')

csv_to_json(datasetName, jsonFileName)
```

```python
indataset = []
with jsonlines.open(jsonFileName) as f:
    for line in f.iter():
        indataset.append(f'<s> {line["inputs"]} </s> {line["responses"]} </s>')

print('데이터셋 확인')
print(indataset[:5])

indataset = Dataset.from_dict({"text": indataset})
indataset.save_to_disk("train_1")

print('데이터셋 info 확인')
print(indataset)
```
