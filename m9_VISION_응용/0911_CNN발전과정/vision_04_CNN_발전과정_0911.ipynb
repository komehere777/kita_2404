## 합성곱 신경망(Convolutional Neural Networks, CNNs)의 발전 과정

초기 단계: Neocognitron
  - 1980년대: CNN의 개념은 1980년대 초반에 후쿠시마 쿠니히코에 의해 소개된 Neocognitron에서 유래합니다. 이는 시각적 패턴을 인식하기 위한 자기조직화 신경망이었으며, 합성곱 계층(conv net)과 풀링 계층의 아이디어를 도입했습니다.

LeNet-5: CNN의 상업적 적용
  - 1998년: 얀 르쿤(Yann LeCun)과 그의 동료들은 손글씨 숫자 인식(mnist)을 위한 CNN 아키텍처인 LeNet-5를 개발했습니다. 이는 실제 응용 프로그램에서 CNN을 사용한 첫 사례 중 하나로, 특히 은행에서 손글씨로 쓴 숫자를 인식하는 데 사용되었습니다.

AlexNet: 딥러닝의 부흥
  - 2012년: 알렉스 크리제브스키, 이야 화이트, 제프리 힌튼이 개발한 AlexNet은 이미지넷 대회(ILSVRC)에서 우승하며 딥러닝의 가능성을 전 세계에 알렸습니다. AlexNet은 깊이, ReLU 활성화 함수, 드롭아웃 등 혁신적인 기술을 통해 이전 모델들을 크게 앞서는 성능을 보였습니다.
VGG, GoogLeNet, ResNet: 깊이와 효율성의 증가
  - 2014년: VGGNet은 단순하고 깊은 아키텍처를 통해 높은 성능을 달성했습니다. 동시에, GoogLeNet(또는 Inception 모델)은 계산 자원을 효율적으로 사용하면서도 성능을 유지하는 방법을 제시했습니다.
  - 2015년: ResNet은 잔차 연결(residual connections)을 도입함으로써 깊은 네트워크의 학습 문제를 해결했습니다.- 기울기 소실문제 이를 통해 100층을 넘어서는 아주 깊은 네트워크를 성공적으로 학습할 수 있게 되었습니다.
이후의 발전
  - 2015년 이후: CNN 아키텍처는 계속해서 발전해 왔습니다. DenseNet, MobileNets, EfficientNet 등은 계산 효율성과 성능을 더욱 향상시켰습니다. 또한, `Attention 메커니즘`- 자연어 처리 뿐아니 이미지도 처리 가능 conv 1d는 시퀀셜한 형태이기 때문 과 같은 새로운 개념이 도입되어, 네트워크가 이미지의 중요한 부분에 집중할 수 있게 되었습니다.
현재와 미래
  - CNN은 이미지 분류, 객체 탐지, 세그멘테이션 등 다양한 컴퓨터 비전 작업에서 핵심적인 역할을 하고 있습니다. 최근에는 자연어 처리(NLP)와 음성 인식 등 다른 분야에도 영향을 미치고 있습니다. 예를 들어 이미지 분류는 이미지 전체를 분석하여 하나의 레이블을 할당하는 작업이라면, `객체 탐지`는 이미지 내에서 여러 객체를 식별하고, 각 객체의 위치(보통 경계 상자(bounding boxes)로 표시)와 클래스를 예측하는 과정입니다.

CNN의 발전 과정은 딥러닝 및 인공지능 분야의 급속한 진전을 상징하며, 계속해서 새로운 연구와 혁신이 이루어지고 있습니다. 이러한 발전은 기술의 한계를 넓히고, 다양한 실제 문제를 해결하는 데 크게 기여하고 있습니다.

### LeNet-5
- 초기 합성곱 신경망(Convolutional Neural Network, CNN) 중 하나로, 1998년에 얀 르쿤(Yann LeCun)과 그의 연구팀에 의해 개발되었습니다. LeNet-5는 손글씨 숫자 인식을 목적으로 만들어졌으며, 특히 미국 우편 서비스에서 손글씨 우편 번호를 자동으로 인식하는 데 사용되었습니다. 이 모델은 딥러닝과 CNN의 초기 성공 사례 중 하나로 꼽히며, 후속 CNN 아키텍처의 발전에 중요한 영향을 미쳤습니다.
- LeNet-5는 복잡하지 않은 망을 이용하여 (당시 기준)높은 성능을 보여주었을 뿐만 아니라 Convolutional layer와 pooling의 조합을 반복하는 현대적인 CNN 구조를 제안했다는 점에서 의미가 있는 모델
- https://velog.io/@woojinn8/CNN-Network-1.-LeNet

LeNet-5의 구조

- 레이어 구성: LeNet-5는 대략적으로 7개의 레이어로 구성되어 있으며, 이 중 3개는 합성곱 레이어(Convolutional layers), 2개는 풀링 레이어(Pooling layers, 당시에는 서브샘플링 레이어라고 불렀음), 그리고 2개는 완전 연결 레이어(Fully connected layers)입니다. 이 레이어들은 손글씨 숫자의 특징을 추출하고, 이를 기반으로 숫자를 분류하는 데 사용됩니다.

- 합성곱과 풀링: LeNet-5는 합성곱 레이어에서 여러 필터를 사용하여 이미지의 특징을 추출합니다. 이후, 풀링 레이어에서는 이 특징의 공간 크기를 줄이며, 변형에 대한 내성을 증가시킵니다.

- 활성화 함수: LeNet-5에서는 시그모이드 함수와 하이퍼볼릭 탄젠트(tanh) 함수를 활성화 함수로 사용했습니다. 이는 당시 널리 사용되던 선택지였으며, 현재는 ReLU(Rectified Linear Unit) 함수 등 더 효율적인 활성화 함수들이 널리 사용되고 있습니다.

- 출력층: 마지막 완전 연결 레이어는 소프트맥스 활성화 함수를 사용하여, 각 클래스(여기서는 0부터 9까지의 숫자)에 대한 확률을 출력합니다. 이를 통해 모델은 주어진 입력 이미지가 어떤 숫자를 나타내는지 예측합니다.

### VGGNet
2014년 옥스퍼드 대학의 Visual Geometry Group에 의해 개발된 합성곱 신경망(Convolutional Neural Network, CNN) 아키텍처입니다. 이 모델은 ImageNet 대회(ILSVRC)에서 뛰어난 성능을 보여주며 광범위한 인기를 얻었습니다. VGGNet의 핵심 특징은 그것의 단순함과 깊이에 있습니다. 이 모델은 여러 층의 작은 크기(3x3)의 필터로 구성된 합성곱 층과, 몇 개의 완전 연결된 층으로 구성되어 있습니다.

[ 주요 특징 ]
- 작은 크기의 필터 사용: VGGNet은 모든 합성곱 층에서 3x3 크기의 필터를 사용합니다. 이 작은 필터 크기를 통해 모델의 깊이를 증가시킬 수 있으며, 이는 복잡한 특징을 더 잘 학습할 수 있게 해줍니다.

- 채널 수의 증가: 네트워크가 깊어질수록 합성곱 층의 채널 수가 두 배씩 증가합니다. 이는 네트워크가 더 복잡한 특징을 학습할 수 있도록 해줍니다.

- 동일한 스트라이드와 패딩 사용: 모든 합성곱 층에서 스트라이드(stride) 1을, 패딩(padding) 'same'을 사용하여 입력과 출력의 공간 차원을 유지합니다.

- 맥스 풀링(Max Pooling) 층: 각 합성곱 층 블록 뒤에는 2x2 크기와 스트라이드 2를 가진 맥스 풀링 층이 있어, 특징 맵의 차원을 줄입니다.

### GoogLeNet, Inception 모듈
- GoogLeNet은 2014년 ImageNet Large-Scale Visual Recognition Challenge (ILSVRC)에서 우승한 심층 합성곱 신경망(Convolutional Neural Network, CNN) 아키텍처입니다. 이 모델은 주로 Inception 모듈이라는 독특한 구성 요소를 통해 인식되며, 효율적인 계산과 깊이(레이어의 수)를 늘리면서도 매개변수의 수를 상대적으로 줄이는 것을 목표로 합니다.
- Inception 모듈의 핵심 아이디어는 네트워크가 스스로 가장 `적합한 합성곱 필터(크기)를 학습`할 수 있게 하는 것입니다. 전통적인 CNN 아키텍처에서는 개발자가 필터의 크기(예: 3x3, 5x5 등)를 사전에 결정해야 합니다. 그러나 Inception 모듈 내에서는 여러 다른 크기의 필터를 병렬로 배치하여, 네트워크가 한 `레이어 내에서 여러 크기의 특징을 동시에 학습`할 수 있도록 합니다. 이는 네트워크가 다양한 스케일의 특징을 더 잘 포착할 수 있게 해줍니다.
- GoogLeNet의 특징
  - GoogLeNet은 여러 개의 Inception 모듈을 포함하며, 깊이가 있으면서도 효율적인 모델을 구현합니다. 모델의 이름은 Google과 LeNet의 조합으로, 이는 구글이 개발에 기여했으며, 초기 CNN 중 하나인 LeNet에 대한 경의를 표하는 의미입니다.
  - 깊이와 너비: GoogLeNet은 매우 깊은 아키텍처를 가지고 있지만, Inception 모듈 덕분에 필요한 매개변수의 수를 크게 줄일 수 있습니다.
  - 계산 효율성: Inception 모듈은 다양한 크기의 필터를 병렬로 사용함으로써, 모델이 한 번의 순전파에서 다양한 크기의 특징을 학습할 수 있게 해줍니다

### ResNet - 잔차 네트워크
ResNet, 또는 Residual Network는 깊은 신경망을 효율적으로 학습시키기 위해 개발된 구조 중 하나로, 2015년에 Microsoft Research의 Kaiming He, Xiangyu Zhang, Shaoqing Ren, 그리고 Jian Sun에 의해 제안되었습니다. ResNet은 그 당시의 깊은 신경망이 직면했던 주요 문제 중 하나인 소실되는 그래디언트(vanishing gradient) 문제를 해결하기 위해 고안되었습니다. 이 아키텍처는 매우 깊은 네트워크를 효과적으로 학습할 수 있도록 하여, 여러 벤치마크에서 뛰어난 성능을 달성했습니다.
- ResNet의 핵심은 잔차 블록(Residual Block)에 있습니다. 각 잔차 블록은 입력을 직접적으로 그 다음 레이어의 출력에 더함으로써, 레이어가 "잔차"를 학습하도록 합니다. 이러한 접근 방식은 레이어가 입력을 그대로 전달하는 것(identical mapping)을 학습하는 것이 레이어가 제로 매핑(zero mapping)을 학습하는 것보다 쉽다는 가정 하에 기반합니다. 이를 통해, 네트워크가 깊어질수록 성능이 향상되는 것을 목표로 합니다.

- ResNet의 특징

깊은 네트워크의 효율적 학습: ResNet은 100층을 넘어서는 매우 깊은 네트워크를 성공적으로 학습할 수 있게 하여, 그 당시 기존의 신경망보다 훨씬 깊은 구조를 가능하게 했습니다.
  - 잔차 학습(Residual Learning): 각 잔차 블록은 입력 특성을 출력에 직접 더함으로써, 레이어가 출력해야 하는 변화량(잔차)만을 학습하게 합니다. 이는 학습 과정을 안정화시키고, 소실되는 그래디언트 문제를 완화합니다.

  - 다양한 변형: ResNet 아키텍처는 그 깊이에 따라 여러 변형이 있습니다. 예를 들어, ResNet-34, ResNet-50, ResNet-101, ResNet-152 등이 있으며, 이 숫자는 네트워크에 있는 레이어의 수를 나타냅니다. ResNet-50 이상의 모델에서는 병목 레이어(bottleneck layer)를 사용하여 계산 효율성을 개선합니다.

  - 전이 학습(Transfer Learning)에의 활용: ResNet은 다양한 시각적 태스크에 대해 사전 학습된 모델을 제공합니다. 이를 통해, 적은 양의 데이터로도 높은 성능의 모델을 개발할 수 있습니다.

## Xception

**Xception**(Extreme Inception)은 Francois Chollet에 의해 제안된 딥러닝 모델로, 특히 **이미지 분류**와 **특징 추출** 작업에서 성능이 뛰어난 네트워크입니다. Xception은 구글이 제안한 **Inception 아키텍처**를 확장한 형태로, Inception 구조의 핵심 아이디어를 더 효율적으로 활용합니다. 이름에서 알 수 있듯이, **"eXtreme Inception"**이라는 의미를 담고 있습니다.

### Xception의 주요 개념과 특징:

1. **Depthwise Separable Convolutions (깊이별 분리 합성곱)**:
   - Xception의 가장 중요한 혁신은 **Depthwise Separable Convolution**을 사용하는 것입니다. 이 방식은 일반적인 합성곱(Convolution)을 두 단계로 나눕니다:
     1. **Depthwise Convolution**: 입력 이미지의 `각 채널`에 대해 독립적으로 합성곱을 수행합니다. 즉, 각 채널에 대해 개별 필터를 적용해 특징을 추출합니다. - 인셉션은 레이어별로 나눠서 적용
     2. **Pointwise Convolution**: 각 픽셀 위치에서 입력의 모든 채널을 선형 결합합니다. 즉, 1x1 크기의 필터를 사용하여 각 위치에서 채널을 결합합니다.

   이 방식은 연산량을 크게 줄이면서도 기존 합성곱 연산의 성능을 유지할 수 있습니다. 결과적으로, **모델의 파라미터 수를 줄이고 학습을 가속화**할 수 있습니다.

2. **Inception과의 차이점**:
   - **Inception**은 다양한 필터 크기를 병렬로 적용해 여러 스케일에서 특징을 추출하는 방식으로, 각 레이어마다 여러 경로(path)를 사용합니다.
   - **Xception**은 이 병렬 구조를 더 간단하게 만들어, 모든 병렬 경로를 독립적인 깊이별 분리 합성곱으로 대체합니다. 이를 통해 더 단순하면서도 성능 좋은 모델을 구현합니다.

3. **효율성과 성능**:
   - Xception은 적은 파라미터로 더 나은 성능을 낼 수 있도록 설계되었습니다. 이는 특히 대규모 데이터셋에서 효과적입니다.
   - Depthwise Separable Convolutions는 연산 효율성을 높여, 적은 연산 비용으로 더 좋은 학습 성능을 보여줍니다.

4. **ImageNet에서의 성능**:
   - Xception은 대규모 이미지 데이터셋인 **ImageNet**에서 높은 정확도를 기록했습니다. 특히, Xception은 Inception V3보다 더 적은 파라미터로 더 나은 성능을 달성했습니다.

5. **네트워크 구조**:
   - Xception 네트워크는 기본적으로 여러 개의 **Depthwise Separable Convolution** 레이어로 구성되어 있습니다. 이는 깊은 네트워크 구조를 가지면서도 연산량과 파라미터 수를 줄여줍니다.
   - 또한 **Residual Connections**(잔차 연결)를 사용해 **정보 흐름을 원활하게 유지**하며, 학습을 안정화합니다. 이는 ResNet에서 제안된 개념으로, 깊은 네트워크에서도 **기울기 소실** 문제를 해결하는 데 도움이 됩니다.

### 요약:
- **Xception**은 **Inception** 아키텍처를 확장하여 **Depthwise Separable Convolution**을 중심으로 설계된 모델로, 적은 연산량과 파라미터로 높은 성능을 보여줍니다.
- **효율적**이고 **강력한 특징 추출**을 가능하게 하며, 이미지 분류나 특징 추출 작업에서 매우 효과적입니다.
- 특히 **대규모 데이터셋**에서 좋은 성능을 보이며, 최신 딥러닝 작업에서 널리 사용되고 있습니다.

```python
from tensorflow.keras.applications.xception import Xception, preprocess_input, decode_predictions

xception = Xception(
    include_top=True,
    weights="imagenet",
    input_tensor=None,
    input_shape=None,
    pooling=None,
    classes=1000,
)

xception.summary()
```

**DenseNet**(Densely Connected Convolutional Networks)은 **CNN(Convolutional Neural Network)**의 한 종류로, 네트워크의 각 레이어가 **이전 모든 레이어의 출력**을 입력으로 받는 구조를 가지고 있습니다. 이는 전통적인 CNN 아키텍처와 달리, **레이어 간의 연결을 더욱 강화**하여 성능을 향상시키는 방식입니다. **DenseNet**은 2017년 Huang et al.에 의해 제안되었으며, 특히 **이미지 분류**, **물체 검출** 및 **세분화** 작업에서 좋은 성능을 발휘합니다.

### DenseNet의 주요 개념과 특징

#### 1. **Dense Connectivity (밀집 연결)**
   - DenseNet의 핵심 아이디어는 **모든 레이어를 서로 연결**하는 것입니다. 각 레이어는 이전 레이어의 출력뿐만 아니라 **그 앞에 있는 모든 레이어의 출력**도 입력으로 받습니다.
   - 예를 들어, 레이어 4는 레이어 1, 2, 3의 출력을 모두 입력으로 받습니다. 이렇게 하면 레이어 간의 정보 흐름이 원활해지며, **특징 재사용**이 가능해집니다.
   
   **Dense Connection**의 장점:
   - **기울기 소실 문제 해결**: DenseNet의 구조 덕분에 역전파(backpropagation) 과정에서 기울기가 소실되지 않고 원활하게 전달될 수 있습니다.
   - **효율적인 특징 학습**: 각 레이어가 이전 레이어에서 이미 학습한 특징을 재사용할 수 있기 때문에 중복되는 정보 학습이 줄어들고, 더 효율적으로 새로운 특징을 학습할 수 있습니다.
   - **파라미터 효율성**: 깊은 네트워크임에도 불구하고 각 레이어가 이전 레이어에서 배운 특징을 활용하기 때문에 파라미터 수가 상대적으로 적습니다.

#### 2. **Dense Block과 Transition Layer**
   - DenseNet은 여러 개의 **Dense Block**으로 구성됩니다. **Dense Block**은 여러 레이어가 밀집 연결된 부분으로, 모든 레이어가 앞의 모든 레이어로부터 입력을 받습니다.
   - **Transition Layer**는 Dense Block 사이에 위치하며, **1x1 합성곱**을 사용해 채널 수를 줄이고, **2x2 평균 풀링**을 사용해 크기를 줄입니다. 이를 통해 네트워크의 복잡도를 줄이고 연산량을 제어할 수 있습니다.

#### 3. **특징 맵의 집적 (Feature Aggregation)**
   - DenseNet은 각 레이어가 독립적인 특징 맵을 학습하지 않고, 앞의 레이어들이 학습한 모든 특징 맵을 **집적하여** 다음 레이어로 전달합니다. 이를 통해 각 레이어가 더욱 복잡한 특징을 학습할 수 있습니다.
   - 이는 일반적인 CNN에서의 레이어별 **특징 학습**을 개선한 것으로, DenseNet에서는 모든 레이어가 전 레이어의 특징을 종합적으로 고려할 수 있습니다.

#### 4. **파라미터 효율성**
   - DenseNet은 **파라미터 수를 크게 줄일 수 있는** 구조를 가지고 있습니다. 레이어들이 서로의 출력을 공유하기 때문에 네트워크 전체의 **필터 수**를 줄여도 성능을 유지할 수 있습니다.
   - 전통적인 네트워크와 비교했을 때, DenseNet은 훨씬 더 적은 파라미터로도 성능을 유지하거나 향상시킬 수 있습니다. 이는 학습 효율성 측면에서 큰 이점입니다.

### DenseNet 아키텍처 구조

DenseNet은 **Dense Block**과 **Transition Layer**의 반복으로 이루어져 있으며, **초기 Convolution Layer**와 **Pooling Layer**, 그리고 마지막에 **Classification Layer**가 있습니다.

1. **Dense Block**:
   - 각 블록은 여러 개의 레이어로 구성되어 있으며, 각 레이어는 이전 모든 레이어의 출력을 입력으로 받습니다. 이로 인해 레이어 간 연결이 매우 밀집됩니다.
   
2. **Transition Layer**:
   - Dense Block 사이에는 Transition Layer가 존재합니다. 이 레이어는 **1x1 합성곱**을 사용해 채널 수를 줄이고, **Pooling**을 통해 공간적 크기를 줄입니다. 이 과정을 통해 네트워크의 복잡도를 제어하고, 연산량을 줄이는 역할을 합니다.
   
3. **Growth Rate**:
   - DenseNet에서는 **Growth Rate**라는 개념이 중요한데, 이는 각 레이어가 출력하는 **특징 맵의 수**를 의미합니다. Growth Rate가 크면 각 레이어가 더 많은 특징을 학습하고, 작으면 상대적으로 적은 특징을 학습합니다.

4. **Bottleneck Layers** (선택적):
   - DenseNet은 파라미터 효율성을 높이기 위해 **Bottleneck Layer**를 사용할 수 있습니다. Bottleneck Layer는 **1x1 합성곱**을 사용하여 채널 수를 줄이고, 이어서 **3x3 합성곱**을 적용하는 방식입니다. 이는 연산량을 줄이면서도 성능을 유지하는 데 도움을 줍니다.

5. **Classification Layer**:
   - 마지막에는 **Global Average Pooling**과 **Fully Connected Layer**를 통해 최종 분류 작업을 수행합니다.

### DenseNet의 주요 장점

1. **기울기 소실 문제 해결**:
   - DenseNet의 밀집 연결 구조 덕분에 정보와 기울기가 네트워크 전반에 잘 전달되어 **기울기 소실 문제**가 줄어듭니다.
   
2. **파라미터 효율성**:
   - 각 레이어가 이전 레이어의 특징을 재사용하기 때문에, 중복된 정보 학습이 줄어들고, 적은 파라미터로도 높은 성능을 낼 수 있습니다.
   
3. **특징 재사용**:
   - DenseNet의 각 레이어는 앞서 학습한 모든 특징을 활용할 수 있기 때문에, 더 **효율적인 특징 학습**이 가능합니다.
   
4. **더 얕은 네트워크로도 좋은 성능**:
   - DenseNet은 일반적인 네트워크보다 더 얕은 구조에서도 좋은 성능을 발휘합니다. 이는 레이어 간 밀집된 연결 덕분에 가능하며, 더 복잡한 네트워크가 필요하지 않다는 장점을 제공합니다.

### DenseNet과 기존 CNN의 차이점

| **특징**              | **DenseNet**                                        | **기존 CNN**                               |
|-----------------------|-----------------------------------------------------|--------------------------------------------|
| **레이어 연결 방식**   | 각 레이어가 모든 이전 레이어의 출력을 입력으로 받음 | 각 레이어는 바로 이전 레이어의 출력만 입력 |
| **기울기 소실 문제**   | 밀집 연결로 기울기 소실 문제가 줄어듦               | 깊어질수록 기울기 소실 문제가 심해짐        |
| **파라미터 효율성**    | 적은 파라미터로 높은 성능                           | 더 많은 파라미터 필요                      |
| **특징 재사용**       | 모든 이전 레이어의 특징을 재사용                    | 이전 레이어의 출력만 사용                  |
| **연산량**            | 상대적으로 적은 연산량                              | 연산량이 많아질 수 있음                    |

### 요약

**DenseNet**은 각 레이어가 이전 레이어의 출력뿐만 아니라 그 앞의 모든 레이어의 출력을 입력으로 받는 **밀집 연결 구조**를 채택한 신경망입니다. 이러한 구조는 **기울기 소실 문제를 해결**하고, **파라미터 효율성을 높이며**, **효율적인 특징 학습**을 가능하게 합니다. DenseNet은 적은 파라미터로도 좋은 성능을 낼 수 있는 모델로, 이미지 분류, 물체 인식 등에서 탁월한 성능을 발휘합니다.

```python
from tensorflow.keras.applications.densenet import DenseNet201, preprocess_input, decode_predictions

densnet = DenseNet201(
    include_top=True,
    weights="imagenet",
    input_tensor=None,
    input_shape=None,
    pooling=None,
    classes=1000,
)

densnet.summary()
```

## 백본(Backbone)
딥러닝과 컴퓨터 비전 분야에서 주로 **특징 추출을 담당하는 주요 신경망 아키텍처**- 프리트레이닝된 모델 를 의미합니다. 특히 이미지 분류, 물체 검출(Object Detection), 세그멘테이션(Segmentation) - 객체의 정확한 형태를 인식 과 같은 다양한 컴퓨터 비전 작업에서 **주요 특징(feature)**을 학습하고 추출하는 역할을 합니다. 백본은 주로 **특징 맵**을 생성해 다른 더 복잡한 모델이나 모듈(예: 물체 검출 헤드, 세그멘테이션 모듈 등)에 입력으로 제공됩니다.

### 백본의 역할

1. **특징 추출**:
   - 백본은 입력 이미지를 받아 **다양한 수준의 특징**을 추출합니다. 저수준 특징(엣지, 패턴 등)부터 고수준 특징(객체의 전체적인 모양 등)까지 다양한 정보를 학습합니다.
   - 예를 들어, 이미지의 **엣지**, **모양**, **패턴**, **텍스처** 등의 정보는 네트워크의 깊이가 깊어짐에 따라 점점 더 추상적인 고수준 정보로 변환됩니다.

2. **사전 학습된 모델 사용**:
   - 백본은 종종 **사전 학습된 네트워크**를 사용합니다. 즉, ImageNet과 같은 대규모 데이터셋에서 미리 학습된 모델을 백본으로 채택하여 특정 작업에 맞게 **전이 학습**(Transfer Learning)을 적용할 수 있습니다. 
   - 이를 통해 **빠른 학습**과 **높은 성능**을 얻을 수 있습니다.

3. **다양한 비전 작업에 적용**:
   - 백본은 여러 컴퓨터 비전 작업에서 공통으로 사용됩니다. 예를 들어, 물체 검출(Object Detection)에서는 백본이 이미지를 받아 물체의 위치와 범주를 추출하는 **RPN(Region Proposal Network)**이나 **FPN(Feature Pyramid Network)** 등의 모듈로 전달합니다.

### 백본으로 사용되는 대표적인 신경망

1. **ResNet (Residual Network)**:
   - ResNet은 물체 인식에서 매우 뛰어난 성능을 보여주는 백본으로, **Residual Connection**을 통해 깊은 네트워크에서도 **기울기 소실** 문제를 해결합니다.
   - 다양한 비전 작업에서 자주 사용되며, ResNet-50, ResNet-101 등의 변형이 존재합니다.

2. **VGG (Visual Geometry Group Network)**:
   - VGG는 깊은 구조와 간단한 **3x3 합성곱 필터**를 사용하는 CNN 아키텍처로, **VGG-16**과 **VGG-19**와 같은 모델들이 백본으로 널리 사용됩니다. 구조가 단순하지만 성능이 뛰어나 여러 작업에서 사용됩니다.

3. **MobileNet**:
   - **경량화된 네트워크**로, **모바일**이나 **임베디드 환경**에서 사용할 수 있도록 설계되었습니다. **Depthwise Separable Convolution**을 사용해 연산량을 줄이고, 작은 기기에서도 사용할 수 있는 백본입니다.

4. **DenseNet**:
   - 각 레이어가 앞서 설명한 것처럼 모든 이전 레이어의 출력을 입력으로 받는 구조로, 밀집 연결을 통해 효율적인 특징 학습을 수행합니다.

5. **Xception**:
   - Inception 아키텍처의 확장형으로, **Depthwise Separable Convolution**을 사용해 효율적으로 특징을 추출합니다. 경량화되면서도 높은 성능을 발휘하는 백본으로, 특히 물체 검출 및 세그멘테이션 작업에 사용됩니다.

### 백본이 적용되는 예시

1. **물체 검출(Object Detection)**:
   - 예를 들어, **Faster R-CNN**이나 **YOLO** 같은 물체 검출 모델에서, 백본은 입력 이미지를 받아 중요한 특징을 추출한 후 **Region Proposal Network (RPN)**이나 **클래스 예측 레이어**에 전달합니다.

2. **이미지 세그멘테이션**:
   - **FCN(Fully Convolutional Network)**, **Mask R-CNN**과 같은 세그멘테이션 모델에서는 백본이 이미지의 각 픽셀에 대한 정보를 추출하여 **세분화** 작업을 수행할 수 있도록 합니다.

3. **이미지 분류(Classification)**:
   - 전통적으로 백본은 이미지 분류 모델에서 먼저 개발되었으며, 입력 이미지를 여러 레이어에 걸쳐 처리하여 클래스 라벨을 예측하는 역할을 합니다.

### 요약

**백본(Backbone)**은 딥러닝 모델에서 **특징 추출을 담당하는 주요 신경망 구조**로, 컴퓨터 비전 작업에서 매우 중요한 역할을 합니다. 이미지의 다양한 특징을 추출하고, 이를 바탕으로 **물체 검출**, **이미지 세그멘테이션**, **이미지 분류** 등의 작업을 수행하는 데 사용됩니다. **ResNet**, **VGG**, **DenseNet**과 같은 모델이 대표적인 백본으로 사용되며, 이를 통해 여러 복잡한 작업에서도 효율적으로 특징을 학습하고 추출할 수 있습니다.
