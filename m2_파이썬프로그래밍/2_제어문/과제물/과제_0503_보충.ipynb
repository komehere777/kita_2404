```python
additional_total = 0
selected_menu = ''
cart_items = 0
cart_total = 0
total_price = 0
total_cart_items = []
set_check = True


def front_page():
    front_page = """
KITA 식당에 오신 것을 환영합니다

========================================

             - 메  뉴 -
     1. 불고기 비빔밥 : 12,000원
     2. 야채  비빔밥  : 8,000원
     3. 전주  비빔밥  : 10,000원

      세트 주문시 : 3000원 추가
   (세트는 밥과 반찬이 추가됩니다.)

========================================"""

    print(front_page)
    cart()


def order():
    # 메뉴별 가격을 설정한다.
    menu_prices = {1:('불고기 비빔밥',12000) , 2:('야채 비빔밥',8000) , 3:('전주 비빔밥',10000) }
    set_prices = {'y':('셋트 포함', 3000), 'n':('', 0)}
    global additional_total
    global selected_menu
    global cart_items
    global cart_total
    global total_price
    global set_check

    while 1:
        # 사용자로부터 메뉴 선택을 받는다.
        menu_choice = int(input("\n비빔밥 메뉴를 선택해 주세요. 1, 2, 3 >>> "))

        # 사용자가 선택한 메뉴의 가격을 가져온다.
        total_price = menu_prices.get(menu_choice)[1]
        selected_menu = menu_prices.get(menu_choice)[0]

        # 세트 주문 여부를 확인한다.
        set_order = input("세트 주문을 하시겠습니까? 3,000원 추가. y, n >>> ").lower()
        if set_order in set_prices:
            if set_order == 'y':
                set_check = True
                total_price += 3000
            else:
                set_check = False
        cart_items += 1
        cart_total += total_price

        cart()
        #추가주문
        additional_order = input("\n추가 주문하시겠습니까? y, n >>> ").lower()
        if additional_order == 'n':
            additional_total += total_price
            print(f"지불하셔야할 총 금액은 {additional_total}입니다.")
            break
        else:
            additional_total += total_price


def cart():
    print("""****************장바구니*****************
    """)
    global selected_menu
    global cart_items
    global cart_total
    global total_cart_items
    global set_check
    total_cart_items.append((selected_menu, set_check))
    print(f"현재 장바구니에 있는 항목은 {cart_items}개입니다.")

    for i in total_cart_items:
        if i == total_cart_items[0]:
            continue
        else:
            print(f"현재 장바구니에 있는 메뉴는 {i[0]}입니다.", end=' ')
            if i[1] == False:
                print("(셋트 미포함)")
            else:
                print("(세트 포함)")
    print(f"\n지금까지 담긴 금액은 {cart_total}입니다.")


def main():
    while True:
        front_page()
        while 1:
            order()
            # 계속 주문하거나 종료 여부를 확인한다.
            continue_ordering = input("\n다시 주문하시겠습니까? (주문: enter, 종료: 'end' 입력) >>> ")
            if continue_ordering == 'end':
                break

        print("\n감사합니다! 다음에 또 방문해 주세요.")
        break


if __name__ == "__main__":
    main()
```

```python
#밀리세컨드에서 랜덤한 숫자1개를 뽑고 영문자 대문자, 소문자 각각3개씩 뽑아 비밀번호를 만든다.
from datetime import datetime
import random
import string

result = ''
a = datetime.now()
mil = list(range(a.microsecond))
ran = random.choice(mil)
upper_str = string.ascii_uppercase
lower_str = string.ascii_lowercase
ran_upper = random.sample(upper_str, 3)
ran_lower = random.sample(lower_str, 3)
word = list(str(ran)) + ran_upper +ran_lower
print(word)
random.shuffle(word)
print(word)
password = ''.join(word)
print("비밀번호는",password)
```

```python
import random

#임의의 숫자로 이루어진 리스트 생성
list1 = []
for i in range(20):
    a = random.randint(1, 100)
    list1.append(a)
print("랜덤 리스트 작성:",list1)
#필터함수를 이용하여 짝수인 리스트만 생성
list_even = list(filter(lambda x : x%2 == 0, list1))
print("람다함수와 필터함을 이용하여 짝수 원소만으로 이루어진 리스트:", list_even)
```
