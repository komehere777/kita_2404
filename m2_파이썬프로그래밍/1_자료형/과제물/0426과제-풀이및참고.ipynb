Task2_0426. 두 개의 리스트에서 공통으로 나타나는 요소들을 찾아 리스트로 반환하는 프로그램을 작성하세요.

예시 입력: list1 = [1, 2, 3, 4, 5], list2 = [4, 5, 6, 7, 8]

예시 출력: [4, 5]

```python
def intersection_list(list1, list2):
    s1 = set(list1)
    s2 = set(list2)
    inter_list = s1 & s2
    return list(inter_list)

list1 = [1, 2, 3, 4, 5]
list2 = [4, 5, 6, 7, 8]

intersection_list(list1, list2)
```

```python
def intersection_list(list1, list2):
    s1 = set(list1)
    s2 = set(list2)
    inter_list = s1.intersection(s2)
    return list(inter_list)

list1 = [1, 2, 3, 4, 5]
list2 = [4, 5, 6, 7, 8]

intersection_list(list1, list2)
```

Task3_0426. 주어진 숫자 리스트에서 특정 값보다 큰 요소가 하나라도 존재하는지 검사하고, 그 결과를 불 값으로 반환하는 함수를 작성하세요.

- numbers = [1, 2, 3, 10, 20]
- threshold = 15

```python
numbers = [1, 2, 3, 10, 20]
threshold = 15

def inspect_num(numbers, threshold):
    for i in numbers:
        if i > threshold:
            return True
    return False

inspect_num(numbers, threshold)
```

Task6_0426. Calculator 클래스를 작성하고 4칙연산을 수행하는 객체 4개를 작성하여 결과를 출력하세요.

```python
class Calculator:
    def setData(self, x, y):
        self.x = x
        self.y = y
    def add(self):
        return self.x + self.y
    def subtract(self):
        return self.x - self.y
    def div(self):
        return self.x / self.y
    def mul(self):
        return self.x * self.y

test1 = Calculator()
test1.setData(10, 2)
print(test1.add())
print(test1.subtract())
print(test1.div())
print(test1.mul())
```

Task7_0426. 다음 과제를 수행하세요.
- 사용자로 부터 텍스트를 입력 받는다.
- 문자을 단어 단위로 분리 : split()
- 단어의 빈도수를 저장할 딕셔너리를 생성
- 각 단어의 빈도 수를 계산(for 문 / if else문)
- 결과 출력

[ 예시 ]

문장을 입력하세요: I love apple. I love orange. Apple is tasty

{'i': 2, 'love': 2, 'apple.': 1, 'orange.': 1, 'apple': 1, 'is': 1, 'tasty': 1}

```python
txt = input("문장을 입력하세요: ").lower()
word = txt.split()

dict = {}
for i in word:
    if i in dict:
        dict[i] += 1
    else:
        dict[i] = 1

print(dict)
```

```python
# 참고
txt = input("문장을 입력하세요: ").lower()


def count_frequent_words(words):
    dict = {}
    words = txt.split()
    for i in words:
        if i in dict:
            dict[i] += 1
        else:
            dict[i] = 1
    return dict

print(count_frequent_words(txt))
```
