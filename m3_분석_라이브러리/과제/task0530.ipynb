Task1_0530. 고객 정보 데이터프레임과 구매 내역 데이터프레임을 결합하여 각 고객이름(name) 별 총 지출을 계산하여 출력하세요.

```python
import pandas as pd

# 고객 정보 데이터프레임 생성
customers = pd.DataFrame({
    'customer_id': [1, 2, 3, 4],
    'name': ['Alice', 'Bob', 'Charlie', 'David']
})

# 구매 내역 데이터프레임 생성
purchases = pd.DataFrame({
    'purchase_id': [101, 102, 103, 104, 105],
    'customer_id': [1, 2, 1, 3, 2],
    'amount': [250, 150, 200, 300, 350]
})
print(customers,'\n')
print(purchases)
#고객이름(name) 별 총 지출을 계산 - 고객이름을 인덱스로
#고객 정보 데이터프레임과 구매 내역 데이터프레임을 결합- 결합 방법은?
#결합방법 - concat, merge(), join()
#커스터머 아이디가 같다 - 같은 열이 있다면? -

#merge()
print("merge")
print(pd.merge(customers, purchases, on='customer_id'))
#join()
# print(customers.join(purchases, on='customer_id'))
print("join")
print(customers.join(purchases, on='customer_id',lsuffix='_1', rsuffix='_2', how='inner'))

merge_df = pd.merge(customers, purchases, on='customer_id')
pivot_merge_df = merge_df.pivot_table(index='name', columns='purchase_id', values='amount', aggfunc='sum', fill_value=0)
print("[결과]")
print(pivot_merge_df)
```

Task2_0530. 학생 성적 데이터프레임과 과목 데이터프레임을 결합하여 각 과목별 평균 성적을 계산하여 출력하세요.

```python
import pandas as pd

# 학생 성적 데이터프레임 생성
grades = pd.DataFrame({
    'student_id': [1, 2, 3, 4, 1, 2, 3, 4],
    'subject_id': [101, 101, 102, 102, 103, 103, 101, 103],
    'grade': [85, 90, 78, 88, 92, 95, 80, 85]
})

# 과목 데이터프레임 생성
subjects = pd.DataFrame({
    'subject_id': [101, 102, 103],
    'subject_name': ['Math', 'Science', 'History']
})
print(grades,'\n')
print(subjects)

merge_df = pd.merge(grades, subjects, on='subject_id')
pivot_merge_df = merge_df.pivot_table(index='subject_name', columns='student_id', values='grade', aggfunc='mean', fill_value=0)
print("[결과]")
print(pivot_merge_df)
```

Task3_0530. 고객 데이터프레임에서 age가 30 이상이고 spending이 500 이상인 고객을 찾아 출력하세요.

```python
import pandas as pd

# 예제 데이터프레임 생성
customers = pd.DataFrame({
    'customer_id': [1, 2, 3, 4, 5],
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'age': [25, 35, 30, 40, 28],
    'spending': [300, 600, 500, 700, 400]
})
customers[(customers['age'] >= 30) & (customers['spending'] >= 500)]
```

Task6_0530. 주어진 titanic3 데이터셋에서 pclass 열을 기준으로 각 객실 등급의 평균 운임(fare)을 계산하여 출력하세요.

```python
df_copy = df.copy()
pivot_df_copy = df_copy.pivot_table(index='pclass', values='fare', aggfunc='mean', fill_value=0    )
print(pivot_df_copy)

df2 = df_copy.groupby(['pclass'])
avg_fare = df2.agg({'fare':'mean'})
print(avg_fare)

df2 = df_copy.groupby(['pclass'])
avg_fare = df2.fare.mean()
print(avg_fare)
```
