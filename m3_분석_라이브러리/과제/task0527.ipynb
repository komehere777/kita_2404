Task1_0527. 4개의 Series를 결합하여 데이터프레임을 생성하여 출력하세요.
- 출력한 데이터프레임에서 추출하고 싶은 5개의 데이터를 인덱싱하여 출력
- null값을 3개 추가
- null값의 개수를 확인하고 삭제
- 2개의 데이터를 수정
- 1개의 행을 삭제

```python
import numpy as np
import pandas as pd

np.random.seed(1)
id = np.arange(1,101)
i1 = pd.Series(id)
korean = np.random.randint(1,101, size=100)
k1 = pd.Series(korean)
math = np.random.randint(1,101, size=100)
m1 = pd.Series(math)
english = np.random.randint(1,101, size=100)
e1 = pd.Series(english)
df = pd.concat([i1,k1,m1,e1], axis=1)
df.rename(columns={0:'id',1:'국어',2:'수학',3:'영어'}, inplace=True)
print(df.head())
print("===========인덱싱================")
print(f'1번 국어점수 :{df.iloc[0,1]}') #id 1, 국어 38?
print(f"5번 영어점수 :{df.loc[4,'영어']}")
print(df.loc[2:3, '국어':'영어'])
print(df.iloc[1],'\n')
print(df.iloc[4],'\n')
print("===========null값 3개 추가============")
df.iloc[0,2] = np.nan
df.iloc[2,2] = np.nan
df.loc[3, '국어'] = np.nan
print(df.head())
print("===========null값 갯수 확인 및 삭제============")
print(df.isnull().sum())
dropped_df = df.dropna()
print(dropped_df.head())
print("===========3 data modify============")
df.iloc[0,2] = 100
df.iloc[2,2] = 100
df.loc[3, '국어'] = 100
print(df.head())
print("===========1 row delete===========")
df1 = df.drop(df.loc[1])
print(df1.head())
```

- S1 ~ S10은 평균 점수를 기준으로 1등급에서 10등급이고 등급간 점수 차는 5점
- 결시자가 국어 3명, 수학 2명 있음
- 영어, 수학의 평균 점수가 국어 대비 5점 낮음

```python
import pandas as pd
import numpy as np
np.random.seed(5)
df = np.random.randint(1,11,size=(10,5))
df=pd.DataFrame(df,index=['s1','s2','s3','s4','s5','s6','s7','s8','s9','s10'],
               columns=['국어','영어','수학','과학','사회'])
#결시자반영
#국어3명
print(df)
df.iloc[0,0] = np.nan
df.iloc[3,0] = np.nan
df.iloc[8,0] = np.nan
print(df)
#수학
df.iloc[0,2] = np.nan
df.iloc[1,2] = np.nan
print(df)
#전체점수 산정
df2 = df*5+50
print(df2)
df2.describe()
#점수 수정
#영어 현재 73점 80점 맞춰야함 평균7점 총점 70점 필요
df2.loc['s3','영어'] = 80
df2.loc['s6','영어'] = 85
df2.loc['s9','영어'] = 80
print(df2)
df2.describe()
#수학
df2.loc['s4','수학'] = 80
df2.loc['s5','수학'] = 80
df2.loc['s7','수학'] = 80
df2.loc['s8','수학'] = 80
df2.loc['s10','수학'] = 80
df2.describe()
```
