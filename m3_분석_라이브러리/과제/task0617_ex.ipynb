Task1_0617. 선그래프, 히스토그램, 산점도, 막대그래프 4개의 subplot으로 구성된 2 x 2 그래프를 작성하세요.
데이터도 그래프에 맞게 생성

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(1)
df = pd.DataFrame({
    'A': np.random.randn(100),
    'B': np.random.randn(100),
    'C': np.random.rand(100) * 100,
    'D': np.random.randint(1, 101, 100)
})

fig, axes = plt.subplots(2, 2, figsize=(10, 6))
df['A'].plot(ax=axes[0,0], title='Line plot')
df['B'].plot(ax=axes[0,1], kind='hist',bins=30 ,title='Histogram')
df.plot(ax=axes[1,0], kind='scatter', x='B', y='C', title='Scatter Plot')
df['D'].head(10).plot(ax=axes[1,1], kind='bar', title='Bar plot')
plt.tight_layout()
plt.show()
```

```python
#ex
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)

ax1.plot(np.arange(20) * 2, (np.arange(20) + np.random.randn(20) * 10), color='blue', marker='o')
ax2.scatter(np.arange(20) * 2, (np.arange(20) + np.random.randn(20) * 10), color='red')
ax3.bar(np.arange(20) * 2, (np.arange(20) + np.random.randn(20) * 10), color='green')
ax4.hist((np.arange(20) + np.random.randn(20) * 10),bins=20, color='yellow')

plt.show
```
