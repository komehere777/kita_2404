Task1_0619. 다음 사항을 수행하세요.
- 첫 번째로 매칭되는 'p' 태그 찾기
- 모든 'p' 태그 찾기
- 클래스가 'content'인 첫 번째 'p' 태그 찾기
- 클래스가 'content'인 모든 'p' 태그 찾기
- 특정 'p' 태그의 모든 부모 태그 찾기
- 특정 'p' 태그의 첫 번째 부모 태그 찾기
- 특정 'p' 태그의 다음 형제 태그 찾기
- 특정 'p' 태그의 이전 형제 태그 찾기
- 특정 'p' 태그 다음에 위치한 모든 태그나 문자열 찾기
- 특정 'p' 태그 이전에 위치한 모든 태그나 문자열 찾기 

```python
from bs4 import BeautifulSoup

html_content = '<html><body><h1>Title</h1><p class="content">First paragraph.</p><p class="content">Second paragraph.</p></body></html>'
soup = BeautifulSoup(html_content, "html.parser")
print(soup.prettify())
```

```python
# - 첫 번째로 매칭되는 'p' 태그 찾기
p_tag = soup.find('p')
print(f'첫번째 p:{p_tag}\n')

# - 모든 'p' 태그 찾기
p_tags = soup.find_all('p')
for i in p_tags:
    print('모든 p: ',i)
print()
# - 클래스가 'content'인 첫 번째 'p' 태그 찾기
con_p = soup.find(class_='content')
print(f'클래스가 content인 첫번째 p: {con_p}\n')
# - 클래스가 'content'인 모든 'p' 태그 찾기
con_ps = soup.find_all(class_='content')
for i in con_ps:
    print("클래스가 content인 모든 p: ",i)
print()
# - 특정 'p' 태그의 모든 부모 태그 찾기
parent_ps = soup.find('p').find_parents()
for i in parent_ps:
    print(i)
print()
# - 특정 'p' 태그의 첫 번째 부모 태그 찾기
parent_p = soup.find('p').find_parent()
print(parent_p,'\n')
# - 특정 'p' 태그의 다음 형제 태그 찾기
next_sib = soup.find('p').find_next_sibling()
print(next_sib,'\n')
# - 특정 'p' 태그의 이전 형제 태그 찾기
pre_sib = soup.find('p').find_previous_sibling()
print(pre_sib,'\n')
# - 특정 'p' 태그 다음에 위치한 모든 태그나 문자열 찾기
a = soup.find('p')
all_next_p = a.find_all_next()
for i in all_next_p:
    print(i)
print()
# - 특정 'p' 태그 이전에 위치한 모든 태그나 문자열 찾기 
all_pre_p = a.find_all_previous()
for i in all_pre_p:
    print(i)
```

Task2_0619. ID를 이용해서 '네이버 뉴스' 추출하세요.

```python
import requests
from bs4 import BeautifulSoup

url = 'https://news.naver.com'
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3',
}
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')
print(soup.prettify())
```

```python
news = soup.select_one('#browserTitleArea')
print(news.text)
```
