Task1_0528. 연산메소드 이용 및 NaN 값을 0으로 대체 후 사칙연산을 수행한 후 결과를 출력하세요.

```python
import pandas as pd
import numpy as np

df1 = pd.DataFrame(np.arange(12.).reshape((3,4)), columns=list('abcd'))
df2 = pd.DataFrame(np.arange(20.).reshape((4,5)), columns=list('abcde'))
df1.loc[1,'b'] = np.nan
print(df1)
print(df2)
ad = df1.add(df2, fill_value=0)
print(ad)
su = df1.sub(df2, fill_value=0)
print(su)
mu = df1.mul(df2, fill_value=0)
print(mu)
di = df1.div(df2, fill_value=0)
print(di)
```

Task2_0528. 주어진 DataFrame의 N열에서 Series s값을 빼고 결과를 새로운 열 O에 저장 후 출력하세요.

```python
import pandas as pd

# 데이터프레임 생성
df = pd.DataFrame({
    "M": [15, 25, 35, 45, 55],
    "N": [100, 200, 300, 400, 500]
})
print(df,'\n')
# Series 생성
s = pd.Series([5, 10, 15, 20, 25])

# series = df[['N']]
series = df.loc[:,'N']
print(series, type(series))
print(s, type(s))
series = series - s
print(series)
df.loc[:,'O'] = series
df
```

Task3_0528. 주어진 DataFrame의 여러 열에 대해 각기 다른 Series를 더하고, 결과를 새로운 DataFrame으로 반환한 후 각 행의 합계를 계산하여 새로운 열에 추가하여 출력하세요.
합계 행 만들~

```python
import pandas as pd

# 데이터프레임 생성
df = pd.DataFrame({
    "A": [1, 2, 3, 4, 5],
    "B": [10, 20, 30, 40, 50],
    "C": [100, 200, 300, 400, 500]
})

# Series 생성
s1 = pd.Series([5, 5, 5, 5, 5])
s2 = pd.Series([10, 10, 10, 10, 10])
s3 = pd.Series([15, 15, 15, 15, 15])

series_a = df.loc[:,'A']
series_a = series_a.add(s1)

print(series_a)
series_b = df.loc[:,'B']
series_b = series_b.add(s2)
print(series_b)
series_c = df.loc[:,'C']
series_c = series_c.add(s3)
print(series_c)
result = series_a.add(series_b).add(series_c)
print(result)
df.loc[:,'합계'] = result
df
```

Task4_0528. df에서 세 열의 값을 더하여 새로운 컬럼을 생성한 후 출력하세요.

```python
import pandas as pd

# 데이터프레임 생성
df = pd.DataFrame({
    "A": [10, 20, 30, 40, 50],
    "B": [1, 2, 3, 4, 5],
    "C": [5, 10, 15, 20, 25],
    "D": [3, 6, 9, 12, 15],
    "E": [2, 4, 6, 8, 10]
})
print(df)
series_a = df.loc[:,'A']
series_b = df.loc[:,'B']
series_c = df.loc[:,'C']
result = series_a.add(series_b).add(series_c)
df.loc[:,'A,B,C합계'] = result
print(df)
```

Task6_0528. df에서 두 개의 열을 더한 값이 다른 한 개의 열보다 큰 경우에는 1, 작은 경우에는 0으로 값을 정하는 새로운 열을 생성한 후 출력하세요.

```python
import pandas as pd

# 데이터프레임 생성
df = pd.DataFrame({
    "A": [2, 3, 4, 5, 6],
    "B": [1, 2, 3, 4, 5],
    "C": [3, 4, 5, 6, 7],
    "D": [8, 9, 10, 11, 12],
    "E": [10, 20, 30, 40, 50]
})
print(df)
s_a = df.loc[:,'A']
s_b = df.loc[:,'B']
s_c = df.loc[:,'C']
s_d = df.loc[:,'D']
s_e = df.loc[:,'E']
sum_ab = s_a.add(s_b)
print(sum_ab)
result_ab_c = sum_ab.sub(s_c)
print(result_ab_c)
decision = lambda x: 1 if x>=0 else 0
res = list(map(decision, result_ab_c))
print(res)
df.loc[:,'AB와 C판정'] = res
sum_bc = s_b.add(s_c)
result_bc_d = sum_bc.sub(s_d)
print(result_bc_d)
res2 = list(map(decision, result_bc_d))
print(res2)
df.loc[:,'BC와 D판정'] = res2
df
```
