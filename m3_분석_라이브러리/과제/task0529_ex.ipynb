Task8_0529. 주어진 DataFrame의 특정 열에 대해 각 요소에 제곱하는 함수를 적용하여 새로운 열을 생성하세요.

```python
import pandas as pd

# 데이터프레임 생성
df = pd.DataFrame({
    "Numbers": [1, 2, 3, 4, 5]
})
# 데이터프레임엔 apply, applymap
df['POW'] = df.apply(lambda x : x * x)
df
```

```python
#ex
df['Numbers'].apply(lambda x : x * x)
df['Numbers'].map(lambda x : x * x)
```

apply vs map
- apply는 DataFrame의 행이나 열 단위로 함수를 적용할 수 있지만, map은 Series의 각 요소에만 함수를 적용
- Series의 각 요소를 변환하는 간단한 작업에는 map이 적합하고, 더 복잡한 변환이 필요한 경우나 DataFrame 전체를 다룰 때는 apply가 유용

```python
import pandas as pd

# 데이터프레임 생성
df = pd.DataFrame({
    "A": [1, 2, 3, 4],
    "B": [5,6,7,8]
})
#행(row)단위로 함수적용(각 행의 합을 계산)
df1 = df.copy()
df1['Row_Sum'] = df1.apply(lambda row: row.sum(), axis=1)
print(df1)
#열(column)단위 함수적용(각 열의 합을 계산)
df2 = df.copy()
df2.loc['Column_Sum'] = df2.apply(lambda row: row.sum(), axis=0)
print(df2)
```

```python
import pandas as pd

# 데이터프레임 생성
df = pd.DataFrame({
    "A": [1, 2, 3, 4],
    "B": [5,6,7,8]
})
#'A'열의 각 요소를 제곱
df['A_Squared'] = df['A'].map(lambda x: x **2)  #시리즈에는 맵
print(df)
```

Task10_0529. 주어진 DataFrame의 각 요소에 대해 10보다 크면 2배, 그렇지 않으면 원래의 수를 반영하는 함수를 적용하여 값을 변환하세요.

- df.apply는 DataFrame의 각 행(row) 또는 열(column)에 함수를 적용합니다. 기본적으로 axis=0으로 설정되어 있어 열 단위로 함수를 적용
- 각 행 또는 열 자체가 Series로 전달되며, 이 Series 전체에 대해 함수를 적용하려고 한다.
- applymap은 DataFrame의 모든 요소에 대해 함수를 적용하므로, 전체 DataFrame에 대한 변환이 필요할 때 유용

```python
import pandas as pd

# 데이터프레임 생성
df = pd.DataFrame({
    "Values": [5, 10, 15, 20, 25]
})

def decision(x):
    if x > 10:
        return x * 2
    else:
        return x
#형태가 시리즈라고 해서 map사용은 안됨
#apply적용하면 에러 발생
df_decision = df.applymap(decision)
print(df_decision)
df_decision2 = df['Values'].map(decision)   #map 사용됨
print(df_decision2)
```

```python
#ex
df['Transformed'] = df['Values'].apply(lambda x: x * 2 if x > 10 else x)
df
```

```python
import pandas as pd

# 예제 데이터프레임 생성
data = {
    'Category': ['A', 'A', 'B', 'B', 'A', 'B', 'B', 'A', 'B', 'A'],
    'Value': [10, 15, 10, 20, 10, 30, 50, 10, 20, 15]
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)
grouped = df.groupby('Category')
print("Grouped DataFrame:")
print(list(grouped))

def z_score(x):
    return (x - x.mean()) / x.std()

#transform 문제점: 트랜스폼이용하면 그룹핑이 없어짐
df_zscore = grouped.transform(z_score)
print(df_zscore)
df['value_zscore'] = grouped.transform(z_score)
print(df)
# df_zscore = grouped.Value.transform(z_score)
# print(df_zscore)
#apply 문제점: 개별 출력은 가능하나 기존 형태로 출력 불가
df_apply_zscore = grouped.apply(z_score)
print(df_apply_zscore)
df['apply_zscore'] = grouped.apply(z_score)
print(df)
```
