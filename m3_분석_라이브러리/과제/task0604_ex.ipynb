Task5. 2020년부터 2025년까지의 분기별 기간 범위를 생성하고, 각 분기의 시작과 끝 날짜를 포함하는 데이터프레임을 작성하세요.

```python

dates = pd.date_range('2020', '2025', freq='Q')
print(len(dates))  # 20행
df = pd.DataFrame(np.random.randn(20, 2),
                  index=dates,
                  columns=['Start','End'])
print(df)
```

```python
#ex
period_range = pd.period_range('2020', '2025', freq='Q')
df = pd.DataFrame({
    'quarter':period_range,
    'start_date':period_range.start_time.date,
    'end_date':period_range.end_time.date
})
print(df)
```

데이터 형식:

date_range는 특정 날짜와 시간을 다루는 DatetimeIndex 객체를 생성.
period_range는 특정 기간을 다루는 PeriodIndex 객체를 생성.

주기 (frequency):

date_range의 주기는 날짜/시간 간격을 나타내며, 이는 Timestamp 객체로 나타남.
period_range의 주기는 특정 기간 간격을 나타내며, 이는 Period 객체로 나타남.

주 사용 사례:

date_range는 시계열 데이터를 생성하고자 할 때 사용 (예: 매일의 주가 데이터).
period_range는 특정 기간 동안의 데이터를 다루고자 할 때 사용 (예: 분기별 매출 데이터).

Task6. 0일부터 10일까지의 시간 범위를 12시간 간격으로 생성하고, 각 간격에 대해 누적 시간을 계산하여 데이터프레임을 작성하세요.

```python
time_range = pd.date_range('2024-06-01','2024-06-11', freq='12h')
print(time_range)
data = {'time': 12}
df = pd.DataFrame(data, index=time_range)
acc_sum = df.resample('12h').sum()
print(acc_sum)
```

```python
# from datetime import timedelta
import pandas as pd

#ex
timedelta_range = pd.timedelta_range(start=0, end='10D', freq='12H')
# print(timedelta_range)
df = pd.DataFrame({
    'timedelta': timedelta_range,
    'cumulative_time':[timedelta.total_seconds() / 3600 for timedelta in timedelta_range] #리스트 컴프리헨션
})
print(df)
```

```python
timedelta_range = pd.timedelta_range(start=0, end='10D',freq='12H')
li = []
for timedelta in timedelta_range:
    li.append(timedelta.total_seconds()/3600)
li
df = pd.DataFrame({
    'timedelta': timedelta_range,
    'cumulative_time': li
})
df
```

Task8. 주어진 df에서 월별로 데이터의 합계를 계산하세요.

```python
#ex
import pandas as pd

# 데이터프레임 생성
df = pd.DataFrame({
    'date': ['2022-01-01', '2022-01-15', '2022-02-01', '2022-02-15', '2022-03-01'],
    'value': [1, 2, 3, 4, 5]
})
df['date'] = pd.to_datetime(df['date'])

df['month'] = df['date'].dt.to_period('M')

monthly_sum = df.groupby('month')['value'].sum().reset_index()
print(monthly_sum)
```

Task9. 주어진 데이터프레임에서 주말(토요일과 일요일) 데이터를 필터링하세요.

```python
import pandas as pd

# 데이터프레임 생성
df = pd.DataFrame({
    'date': ['2022-01-01', '2022-01-02', '2022-01-03', '2022-01-08', '2022-01-09'],
    'value': [1, 2, 3, 4, 5]
})
df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)
print(df)
dates = pd.date_range('2022-01-01','2022-01-09', freq='B')
print(dates)
week_day_df = pd.DataFrame(columns=['value'], index=dates)
print(week_day_df)
res = pd.merge(df, week_day_df, how='outer', indicator=True)
print(res)
res = res.query("_merge=='left_only'")
res = res.drop(columns=['_merge'])
```

```python
#ex
import pandas as pd

# 데이터프레임 생성
df = pd.DataFrame({
    'date': ['2022-01-01', '2022-01-02', '2022-01-03', '2022-01-08', '2022-01-09'],
    'value': [1, 2, 3, 4, 5]
})
df['date'] = pd.to_datetime(df['date'])

#주말 데이터 필터링
df['day_of_week'] = df['date'].dt.dayofweek
#주말이 5,6
weekend_df = df[df['day_of_week'] >= 5]
weekend_df
```
