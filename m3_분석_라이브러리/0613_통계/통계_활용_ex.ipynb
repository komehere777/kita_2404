Q. 한 지역의 주택 가격 데이터가 다음과 같습니다:

[250000, 300000, 350000, 400000, 450000, 500000, 550000, 600000, 650000, 700000].

주택 가격의 분포를 시각화하기 위해 분포 그래프를 그리고, 주택 가격 분포에 대한 인사이트를 도출하세요.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
house = pd.DataFrame([250000, 300000, 350000, 400000, 450000, 500000, 550000, 600000, 650000, 700000])

house.plot(kind='hist')
#해당 지역은 각 가격대 별로 균일하게 물건이 분포하고 있어
#모든 수요자군을 충족시키고 있다고 보인다.
#어느 특정 수요자군을 해당 지역으로 전입할 경우 공급 부족으로 인해
#특정 가격대 물건의 수요가 증가하여 다른 가격대의 물건의 가격을 넘을 확률이 존재한다.
```

```python
#ex

house = np.array([250000, 300000, 350000, 400000, 450000, 500000, 550000, 600000, 650000, 700000])

#밀도그래프 그리기
sns.kdeplot(house, fill=True)
#정규분포 형태가 분석하기 좋음, 왜곡되어 있으면 분석하기 어려움
#인사이트 해석
mean_price = np.mean(house)
std_price = np.std(house)
median_price = np.median(house)
print(mean_price)
print(std_price)
print(median_price)
#평균과 중앙값이 동일, 고르게 분포
if mean_price > median_price:
    print('주택 가격 분포가 오른쪽으로 치우쳐 있음을 의미합니다.')
elif mean_price < median_price:
    print('주택 가격 분포가 왼쪽으로 치우쳐 있음을 의미합니다.')
else:
    print('주택 가격이 고르게 분포되어 있음을 의미합니다.')
```

Q. 한 회사의 연구개발(R&D) 투자 비용과 연간 수익 데이터가 각각 다음과 같습니다:

투자 비용 [100, 200, 300, 400, 500] (단위: 천 달러),

연간 수익 [200, 400, 600, 800, 1000] (단위: 천 달러).

추세선이 있는 산점도를 그려 투자 비용과 연간 수익 간의 관계를 시각화하고, 인사이트를 도출하세요.

```python
rnd = pd.DataFrame([100, 200, 300, 400, 500])
income = pd.DataFrame([200, 400, 600, 800, 1000])
plt.plot(rnd, income, '-o')
plt.show()
#추세선있는 산점도를 볼경우 투자비용과 연간 수익은 비례관계에 있다.
#해당 연간 수익이 투자비용과 완벽한 관련이 있다는 것이 판명될 경우
#투자 비용을 계속 늘려 연간 수익이 증가함을 판단하면 되겠다.
```

```python
#ex

rnd = np.array([100, 200, 300, 400, 500])
annual = np.array([150, 400, 660, 900, 1000])
#오차가 최소화하는 선을 그리면 추세선이 나옴

plt.scatter(rnd, annual)
m, b = np.polyfit(rnd, annual, 1)
plt.plot(rnd, m * rnd + b, 'r')
plt.xlabel('rnd')
plt.ylabel('annual')
plt.show()

correlation = np.corrcoef(rnd, annual)[0,1]
print(correlation)

if correlation > 0.7:
    print('strong postive correlation')
elif correlation < -0.7:
    print('strong minus correlation')
else:
    print('약하거나 없음')
#
```
