## 통계분석과 기계학습

통계분석
- 통계분석은 데이터로부터 모델을 만들어 `현상을 설명하고 이해`하는 데 중점 - *기술통계에서 다룸, 가설에 대한 검증과 판단을 하는 방법도 사용*


- 단순하고 해석 가능한 형태. 이 모델들은 명확한 `가정`(예: 정규성, 독립성, 등분산성)에 기반하고, 이러한 가정들은 모델의 유효성을 판단하는 데 중요 - *딥러닝은 선형변환과 비선형변환도 사용, 통계에서는 비선형변환이 설명안됨, 회귀분석은 통계분석방법, 트리모델로 가면 통계와 다름, 통계학과 컴공과 관심 포인트가 다름, 컴공은 가정은 불필요 결과만 나오면 됨, 통계는 인과관계 중심, 현실 데이터가 정규성을 가지고 있지 않음, 머신러닝을 통해 패턴만 발견하면 됨, 통계란 데이터 정제에 관점에서 접근, 패턴만 잘 발견되면 됨, 딥러닝은 컴피티션으로 메인되는 방법이 계속 진화, 성능 높은게 짱, 논문에는 기존대비 얼마 성능 업에 대한 내용이 있음*
- 통계분석의 결과는 데이터에 대한 해석과 추론에 중점

기계학습
-  데이터로부터 패턴을 학습하여 미래 데이터에 대한 `예측을 하는 데 중점`
-  복잡하고, 통계적 가정보다는 `예측 성능에 초점`을 둔다. 이러한 모델들은 때로 해석하기 어려울 수 있으며, `"블랙 박스"`로 간주 - 대책 일반 인공지능의 위험성에 대해 관심, 로직을 알아야하는것 아닌가? 결과와 과정이 대립, 안을 들여다 볼려는 시도가 있음
- 기계학습의 결과는 주로 예측의 `정확도와 모델의 성능에 중점`. 해석 가능성보다는 예측 성능이 더 중요한 역할, 평가 방법이 중요함, 평가 툴도 많음

머신러닝과 인공지능에는 여러 가지 통계적 개념과 기법들이 필요
- 통계가 필요한 이유는 정제된 데이터를 넣기 위함!

1. 기초 통계 개념
- 기술 통계 (Descriptive Statistics): `데이터의 기본적인 특성을 요약하고 설명`합니다. 평균, 중앙값, 표준편차, 분산, 범위, 사분위수 등이 포함.
- 추론 통계 (Inferential Statistics): 표본 데이터를 이용해 모집단에 대한 결론을 도출. 가설 검정, 신뢰 구간, p-value 등이 포함. - 결과와 데이터간 관계를 아는 것이 중요, 범주형 데이터들은 어떻게 해석할것인가, 종속변수와 관계가 유의미한가?, age란 변수가 survived와의 관계 age는 범주형 데이터, 범주형은 상관관계가 잘 안나옴, 분산을 가지고 보는게 있음 (카이제곱)
2. 확률 이론 (Probability Theory) - 중요!!, 전체 집단을 다 다룰수없어 표본으로 해석, 랜덤하게 잘 뽑으면 표본에서도 그 관계를 파악가능, 그래서 확률이 중요
- 확률 분포 (Probability Distributions): 데이터가 어떤 패턴을 따르는지 설명합니다- 분포에 맞는 방법을 선택하기 위해 패턴을 이해가 필요. 데이터가 많으면 많을 수록 정규분포를 가진다고 가정함. 정규분포, 이항분포, 포아송분포 등이 포함.
- 확률 변수 (Random Variables): 결과가 확률에 따라 결정되는 변수- 주사위 던지기 이벤트에서는 확률변수(1~6숫자), 1/6이 확률. 이산 확률 변수(끊어져서 나오는 경우 주사위)와 연속 확률 변수로 나눌 수 있다.
- 기대값(평균으로 볼수있음)과 분산(평균으로부터 얼마나 퍼져있는가?,변동성)  (Expected Value and Variance): 확률 변수의 평균과 변동성을 측정.
3. 가설 검정 (Hypothesis Testing)- 통계에서 많이 쓰나 머신러닝에서 는 많이 안씀
- 귀무가설과 대립가설 (Null and Alternative Hypothesis): 검정의 기본 가정. 귀무가설은 보통 '효과 없음'을 가정.
- t-검정 (t-test): 두 그룹의 평균을 비교하는 데 사용.
- 카이제곱 검정 (Chi-Square Test): 범주형 데이터의 분포를 비교.
- ANOVA (Analysis of Variance): 세 개 이상의 그룹 간의 평균 차이를 비교.
4. 회귀 분석 (Regression Analysis) - 머신러닝과 유사, 기울기와 절편 구하기 머신러닝과 유사
- 단순 선형 회귀 (Simple Linear Regression): 하나의 독립 변수와 종속 변수 간의 관계를 모델링.
- 다중 선형 회귀 (Multiple Linear Regression): 여러 독립 변수와 종속 변수 간의 관계를 모델링.
- 로지스틱 회귀 (Logistic Regression): 종속 변수가 범주형일 때 사용.회귀를 분류모델로 바꾸기 위해 적용
5. 확률 모델링 (Probabilistic Modeling)
- 베이지안 통계 (Bayesian Statistics): 사전 지식과 데이터를 결합하여 결론을 도출.머신러닝에서 많이 사용, 어떤 일이 있을것을 조건으로 설명
6. 시간 시계열 분석 (Time Series Analysis)
- 추세 분석 (Trend Analysis): 데이터의 장기적인 증가 또는 감소 경향을 파악.
- 계절성 분석 (Seasonality Analysis): 데이터의 주기적인 패턴을 분석.
- ARIMA 모델 (AutoRegressive Integrated Moving Average): 시간 시계열 데이터를 분석하고 예측하는 데 사용.
7. 고급 통계 기법
- 주성분 분석 (PCA, Principal Component Analysis): 고차원의 데이터를 저차원으로 축소하여 중요한 패턴을 찾는다.
- 클러스터링 (Clustering): 데이터 포인트를 유사한 그룹으로 나눈다. K-평균, 계층적 클러스터링 등이 포함.

# 기초통계

평균과 표준편차

데이터의 중심 경향성과 분산을 파악하는 데 유용. 예를 들어, 학생들의 시험 점수 데이터를 분석할 때 평균과 표준편차를 통해 전체 성적 수준과 변동성을 이해할 수 있다.

```python
import numpy as np

data = np.array([85,90,78, 92, 88, 76, 95, 89, 77, 91])

mean = np.mean(data)
std = np.std(data)

print("평균:", mean)
print("표준편차:", std)
```

중간값과 사분위수

데이터의 중심 경향과 분포를 이해하는 데 도움을 준다. 예를 들어, 가계 소득 데이터를 분석할 때 중간값과 사분위수를 통해 소득 분포를 파악할 수 있다.

어느쪽으로 쏠려있는지 파악가능,평균과 분산으로는 파악 어려움

```python
median = np.median(data)
q1 = np.percentile(data, 25) #25%
q3 = np.percentile(data, 75)

print("중간값:", median)
print("1사분위수:", q1)
print("3사분위수:", q3)
```

평균값 > 중앙값
- 데이터는 오른쪽으로 치우친(오른쪽 꼬리가 긴) 분포를 가진다. 이를 '오른쪽으로 치우친 분포' 또는 '양의 왜곡 분포'라고 한다.
- 특징: 데이터의 높은 값(오른쪽 꼬리)에 이상치나 극단적인 높은 값들이 존재한다. 그것이 평균 끌어올림
- 예시: 적은 수의 매우 높은 소득을 가진 사람들이 있는 소득 데이터. 이러한 극단적으로 높은 값들은 평균을 중앙값보다 더 높게 밀어올린다.

중앙값 > 평균값
- 데이터는 일반적으로 왼쪽으로 치우친(왼쪽 꼬리가 긴) 분포를 가진다. 이러한 분포를 '왼쪽으로 치우친 분포' 또는 '음의 왜곡 분포'라고 한다.
- 특징: 데이터의 낮은 값(왼쪽 꼬리)에 이상치나 극단적인 낮은 값들이 존재할 수 있다.
- 예시: 적은 수의 매우 낮은 소득을 가진 사람들이 있는 소득 데이터. 이러한 극단적으로 낮은 값들은 평균을 중앙값보다 더 낮게 끌어내린다.

중요성

평균과 중앙값 사이의 차이는 데이터의 비대칭성을 나타내며 이 차이가 클수록 데이터 분포의 왜곡이 더 크다고 볼 수 있다. 데이터 분석에서는 이러한 왜곡을 인지하고 적절한 해석과 분석 방법을 선택하는 것이 중요. 예를 들어, 이상치나 극단적인 값이 데이터의 해석에 큰 영향을 미칠 수 있으므로, 이를 고려하여 분석을 진행. - 학습에도 영향, 모델의 성능이 떨어짐, 그래서 과대적합을 방지, 과대적합을 하면 소수의 이상치도 학습됨, 릿지, 랏소등은 이상치 요소를 모델에 영향을 주지 않는 방법을 적용, 이런걸 알기 위해 통계에 대한 이해가 필요

치우친 정도로  분포가 예측 가능

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 데이터 생성
data_right_skewed = np.random.beta(a=2, b=10, size=1000)  # 오른쪽으로 긴 꼬리
data_left_skewed = np.random.beta(a=10, b=2, size=1000)   # 왼쪽으로 긴 꼬리

# DataFrame 생성
df_right_skewed = pd.DataFrame(data_right_skewed, columns=['val'])
df_left_skewed = pd.DataFrame(data_left_skewed, columns=['val'])

# 히스토그램 그리기
fig, axs = plt.subplots(1, 2, figsize=(15, 5))

# 오른쪽으로 긴 꼬리 분포
axs[0].hist(df_right_skewed['val'], bins=30, color='blue', alpha=0.7)
axs[0].axvline(np.mean(df_right_skewed['val']), color='red', linestyle='dashed', linewidth=1)
axs[0].axvline(np.median(df_right_skewed['val']), color='green', linestyle='dashed', linewidth=1)
axs[0].legend({'Mean':np.mean(df_right_skewed['val']), 'Median':np.median(df_right_skewed['val'])})
axs[0].set_title('Right-Skewed Distribution')

# 왼쪽으로 긴 꼬리 분포
axs[1].hist(df_left_skewed['val'], bins=30, color='orange', alpha=0.7)
axs[1].axvline(np.mean(df_left_skewed['val']), color='red', linestyle='dashed', linewidth=1)
axs[1].axvline(np.median(df_left_skewed['val']), color='green', linestyle='dashed', linewidth=1)
axs[1].legend({'Mean':np.mean(df_left_skewed['val']), 'Median':np.median(df_left_skewed['val'])})
axs[1].set_title('Left-Skewed Distribution')

plt.tight_layout()
plt.show()


```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(0)
ar1 = np.random.normal(50, 10, 100)
ar1_median = np.median(ar1)
ar1_mean = np.mean(ar1)

ar2 = ar1.copy()
ar2[0] = 500
ar2_median = np.median(ar2)
ar2_mean = np.mean(ar2)

# 중위수와 평균 차이 출력
print(ar2_median - ar1_median)
print(ar2_mean - ar1_mean)
```

의미와 중요성

- 이상치의 영향: 평균값은 이상치에 의해 크게 왜곡될 수 있다. 예를 들어, 소득 데이터에서 몇몇 매우 높은 값(억만장자의 소득)이 평균을 크게 올릴 수 있지만, 중앙값은 이러한 극단적인 값의 영향을 덜 받는다. - 극단적인 값이 있는 경우 중앙값 이용
- 데이터 분포의 대표성: 중앙값은 데이터의 분포가 비대칭일 때 데이터의 "중심"을 더 잘 나타내는 경향이 있다. 반면, 평균값은 데이터가 대칭적인 분포를 가질 때 중심 경향을 잘 나타낸다.
- 데이터 해석: 데이터 분석에서는 종종 평균값과 중앙값을 함께 사용하여 데이터의 분포를 더 잘 이해한다. `두 척도 사이에 큰 차이가 있다면, 이는 데이터에 이상치나 극단적인 값이 있다는 신호`일 수 있다.
따라서 평균값과 중앙값을 선택할 때는 데이터의 특성과 분석의 목적을 고려해야 하며 이상치가 존재하거나 데이터가 비대칭적인 분포를 가질 경우, 중앙값이 더 적절할 수 있다. 반면, 데이터가 정규 분포를 이루는 경우, 평균값이 데이터를 대표하는 데 적합할 수 있다.

확률(Probability)

확률은 어떤 사건이 일어날 가능성을 수치로 표현한 것입니다. 확률 값은 0에서 1 사이의 숫자로 표현되며, 0은 불가능한 사건, 1은 확실한 사건을 나타냅니다.
확률의 기본 성질:
- $0≤P(A)≤1$
- 모든 가능한 사건들의 확률의 합은 1이다. $(∑P(Ai)=1)$

확률변수(Random Variable)
- 확률변수는 무작위 실험의 결과를 수치로 나타내는 변수입니다. 확률변수는 이산형(discrete)과 연속형(continuous)으로 나눌 수 있습니다.
- 예를 들어, 주사위를 던질 때 나오는 눈의 수는 이산형 확률변수이고, 특정 시간 동안의 온도는 연속형 확률변수입니다.

확률분포(Probability Distribution)
- 확률분포는 확률변수가 취할 수 있는 값과 그 값이 나타날 확률을 나타낸 것입니다.
이산형 확률변수의 경우, 확률질량함수(PMF: Probability Mass Function)를 사용하여 각 값이 나타날 확률을 나타냅니다.
- 연속형 확률변수의 경우, 확률밀도함수(PDF: Probability Density Function)를 사용하여 특정 `구간` 내에서 값이 나타날 확률을 나타냅니다.
- 누적분포함수(CDF: Cumulative Distribution Function)는 확률변수가 특정 값 이하일 확률을 나타냅니다.

```python
#주사위를 두번 던질때, 두눈의 합이 7이 될 확률
#(product(dice, repeat=2)):다이스 리스트에서 2개의 요소를 뽑아 순서를 고려한 모든 가능한 조합 생성
#동일한 iterable을 여러 번 사용 가능하여 순열과는 차이가 있음 - 데카르트의 곱

from itertools import product

dice = [1, 2, 3, 4, 5, 6]
outcomes = list(product(dice, repeat=2))

favorable_outcomes = [(a, b) for a, b in outcomes if a + b ==7]
probability = len(favorable_outcomes) / len(outcomes)

print("favorable outcomes:", favorable_outcomes, len(favorable_outcomes))
print('outcomes',len(outcomes),outcomes )
print("probability:", probability)
```

```python
# 정규분포를 따르는 데이터에서 평균이 0, 표준편차가 1일 때, 값이 1보다 작을 확률
#누적분포함수(CDF: Cumulative Distribution Function)는 확률변수가 특정 값 이하일 확률을 나타냅니다.
#누적되는 확률로 생각
import scipy.stats as stats

mean = 0
std_dev = 1
value = 1

probability = stats.norm.cdf(value, mean, std_dev)

print("probability:", probability)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# 정규분포의 평균과 표준편차
mean = 0
std_dev = 1
value = 1

# 특정 값에 대한 누적 확률 계산
probability = stats.norm.cdf(value, mean, std_dev)

# 정규분포 곡선 데이터 생성
x = np.linspace(-3, 3, 1000)
y = stats.norm.pdf(x, mean, std_dev)

# 시각화
plt.figure(figsize=(10, 6))

# 정규분포 곡선
plt.plot(x, y, label='Normal Distribution', color='blue')

# 특정 값까지의 영역 색칠
x_fill = np.linspace(-3, value, 1000)
y_fill = stats.norm.pdf(x_fill, mean, std_dev)
plt.fill_between(x_fill, y_fill, alpha=0.3, color='blue', label=f'Cumulative Probability for value <= {value}')

# 그래프 설정
plt.title('Normal Distribution and Cumulative Probability')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.axvline(value, color='red', linestyle='--', label=f'Value = {value}')
plt.legend()

# 확률 값 출력
plt.text(-2, 0.3, f'P(X <= {value}) = {probability:.2f}', fontsize=12)

plt.show()
```

```python
#이산형
#n총 시행횟수, k성공횟수, p성공할 확률(동전은 1/2확률)
#이산형 확률변수의 경우, 확률질량함수(PMF: Probability Mass Function)를 사용하여 각 값이 나타날 확률을 나타냅니다.
#동전을 10번 던질 때, 앞면이 정확히 6번 나올 확률

from scipy.stats import binom

n = 10 #총 시행횟수
k = 6 #성공횟수
p = 0.5 #성공할 확률

probability = binom.pmf(k, n, p)

print("probability:", probability)
```

기대값(Expectation Value)

- 확률분포의 평균을 의미하며, 확률변수가 가질 수 있는 값들에 대한 확률 `가중 평균`

```python
[1/6] * 6
```

```python
#주사위를 던질 때 나오는 눈의 수의 기대값(확률의 가중 평균)
values = [1,2,3,4,5,6]
probabilities = [1/6] * 6  #이곳 확률 비율이 달라지면 기대값이 달라짐
#기대값 계산
expected_value = sum(value * probability for value, probability in zip(values, probabilities))

print("expected value:", expected_value)
```

**분산(Variance)**
- 확률변수가 평균으로부터 얼마나 떨어져 있는지를 나타내는 지표입니다.
- 분산은 확률분포의 흩어짐 정도를 나타내며, 기대값이 중심에 있을 때 개별 데이터가 얼마나 퍼져 있는지를 보여줍니다.
- 분산이 클수록 데이터가 평균값 주위에 널리 퍼져 있고, 분산이 작을수록 데이터가 평균값에 더 가깝게 모여 있습니다.

분산의 공식은 다음과 같습니다:

- 이산형 확률변수의 분산: $σ2 =∑(xi −μ)2 P(xi)$
- 연속형 확률변수의 분산(적분, 범위를 따져야하니): $σ2 =∫(x−μ)2f(x)dx$

여기서 $μ$는 기대값(평균), $xi$는 확률변수가 가질 수 있는 값, $P(xi)$는 각 값이 나타날 확률입니다.

주사위를 던질 때 나오는 눈의 수의 분산
- 주사위 눈의 분산을 계산하여 주사위 결과가 얼마나 흩어져 있는지를 이해
- 분산이 클수록 주사위 결과가 널리 퍼져 있음을 의미

```python
#주사위를 던질 때 나오는 눈의 수의 기대값(확률의 가중 평균)
values = [1,2,3,4,5,6]
probabilities = [1/6] * 6  #이곳 확률 비율이 달라지면 기대값이 달라짐
#기대값 계산
expected_value = sum(value * probability for value, probability in zip(values, probabilities))

print("expected value:", expected_value)
#분산의 제곱의 합
#표준편차는 분산에 루트

#주사위 눈의 분산
variance = sum((value - expected_value)**2 * probability for value, probability in zip(values, probabilities))
print("variance:", variance)
```

어떤 시험에서 점수가{60,70,80,90,100}로 구성될때 각 점수의 확률은 {0.1,0.2,0.4,0.2,0.1}로 주어집니다.
분산을 계산하여 점수가 얼마나 다양한지 파악

```python
scores = [60,70,80,90,100]
probabilities = [0.1,0.2,0.4,0.2,0.1]
expected_value = sum(score * probability for score, probability in zip(scores, probabilities))

print("expected value:", expected_value)
#분산의 제곱의 합
#표준편차는 분산에 루트

#주사위 눈의 분산
variance = sum((score - expected_value)**2 * probability for score, probability in zip(scores, probabilities))
print("variance:", variance)

#표준편차는
std = variance ** 0.5
print(std)
```

```python
#스코어를 중괄호로 할 경우 분산이 달리 나오는 이유는 set이기 때문 순서가 없이 들어감
#확률이 일정하지 않아서 가중 평균이 순서에 따라 달리 나옴
scores = {60,70,80,90,100}
probabilities = [0.1,0.2,0.4,0.2,0.1]
expected_value = sum(score * probability for score, probability in zip(scores, probabilities))

print("expected value:", expected_value)
#분산의 제곱의 합
#표준편차는 분산에 루트

#주사위 눈의 분산
variance = sum((score - expected_value)**2 * probability for score, probability in zip(scores, probabilities))
print("variance:", variance)

#표준편차는
std = variance ** 0.5
print(std)
```
