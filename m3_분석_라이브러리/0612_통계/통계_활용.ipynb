Q. 한 학급의 학생 10명의 수학 시험 점수는 다음과 같습니다:

[85, 72, 90, 68, 88, 76, 95, 89, 77, 84].

학급의 평균 점수와 표준편차를 계산하여 전체 성적 수준과 변동성을 파악하세요. 이 데이터에서 어떤 인사이트를 얻을 수 있습니까?

```python
import pandas as pd
import numpy as np
scores = pd.DataFrame([85, 72, 90, 68, 88, 76, 95, 89, 77, 84])
print(scores.mean())
print(scores.std())
print(scores.describe())
#평균은 82.4점이고 ,표준편차는 8.28로 최저 점수는 평균과 14점차이고 최고점수는 13점 차이이므로 균일하게 분포한것 같다.
```

Q. 한 식당의 월별 홍보 비용과 월별 방문자 수 데이터가 각각 다음과 같습니다:

홍보 비용 [500, 600, 700, 800, 900] (단위: 달러),

방문자 수 [300, 400, 500, 600, 700] (단위: 명).

홍보 비용과 방문자 수 간의 상관 계수를 계산하여 관계를 파악하세요. 이 데이터에서 어떤 인사이트를 얻을 수 있습니까?

```python
ad = pd.Series( [500, 600, 700, 800, 900])
customers = pd.Series([300, 400, 500, 600, 700] )

ad_cus_df = pd.concat([ad, customers], axis=1)
print(ad_cus_df)
print(ad.corr(customers))
#0.7 이상으로 강한 상관관계에 있어 홍보비용 지출은 방문자수와 관계가 있어
#자본의 여력이 되는 한 홍보비용 지출이 요구된다.
```

Q. 한 회사의 연구개발(R&D) 투자 비용과 연간 수익 데이터가 각각 다음과 같습니다:

투자 비용 [100, 200, 300, 400, 500] (단위: 천 달러),

연간 수익 [200, 400, 600, 800, 1000] (단위: 천 달러).

추세선이 있는 산점도를 그려 투자 비용과 연간 수익 간의 관계를 시각화하고, 인사이트를 도출하세요.

```python
rnd = pd.DataFrame([100, 200, 300, 400, 500])
income = pd.DataFrame([200, 400, 600, 800, 1000])
plt.plot(rnd, income, '-o')
plt.show()
#추세선있는 산점도를 볼경우 투자비용과 연간 수익은 비례관계에 있다.
#해당 연간 수익이 투자비용과 완벽한 관련이 있다는 것이 판명될 경우
#투자 비용을 계속 늘려 연간 수익이 증가함을 판단하면 되겠다.
```
