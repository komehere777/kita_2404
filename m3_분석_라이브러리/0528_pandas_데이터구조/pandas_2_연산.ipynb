## 기초 **연산**

```python
import pandas as pd

st1 = pd.Series({'국어':100 ,'영어':80 , '수학':90})
st2 = pd.Series({'국어':90 ,'영어':80 , '수학':80})

print(st1, '\n')
print(st2)
```

```python
pd.DataFrame(st1+st2)
```

```python
#두 학생의 과목별 점수로 사칙연산 수행
add = st1 + st2
sub = st1 - st2
mul = st1 * st2
div = round((st1 / st2),2)

df = pd.concat([add,sub,mul,div],axis=1)
print(df)
```

```python
df = pd.concat([add,sub,mul,div],axis=1)
df.columns = ['덧셈','뺄셈','곱셈','나눗셈']
df
```

```python
print(add,type(add))    #시리즈
```

```python
result = pd.DataFrame([add,sub,mul,div])               #데이터프레임에 리스트로 넣으면 행으로 들어감
result
```

```python
result = pd.DataFrame([add,sub,mul,div],index=['덧셈','뺄셈','곱셈','나눗셈'])               #인덱스로 넣줘야함
result
```

```python
import numpy as np

st1 = pd.Series({'국어':np.nan,'영어':80,'수학':90})
st2 = pd.Series({'수학':80,'국어':90})

add = st1 + st2
sub = st1 - st2
mul = st1 * st2
div = round((st1 / st2),2)

result = pd.DataFrame([add,sub,mul,div],
                     index=['덧셈','뺄셈','곱셈','나눗셈'])
result
```

```python
#Q.null값을 0으로 변경하여 결과를 출력하세요
st1 = pd.Series({'국어':np.nan,'영어':80,'수학':90})
st2 = pd.Series({'수학':80,'국어':90})

ad = st1.add(st2, fill_value=0)
su = st1.sub(st2, fill_value=0)
mu = st1.mul(st2, fill_value=0)
di = st1.div(st2, fill_value=0)
result = pd.DataFrame([ad,su,mu,di],
                     index=['덧셈','뺄셈','곱셈','나눗셈'])
result
```
