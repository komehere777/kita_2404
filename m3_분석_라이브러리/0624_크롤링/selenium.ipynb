````python
#implicitly_wait() : 웹 페이지에서 사용되는 모든 요소들을 로드할 때까지 대기
#By class는 요소를 찾기 위한 다양한 방법을 제공 
#webDriverWait class는 특정 조건이 만족할 때까지 대기하는 기능을 제공
#expected_conditions class는 웹 페이지의 특정 요소가 로드되는 조건을 지정 특정 상태가 될때가지 기다리는 기능을 제공

from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

driver = webdriver.Chrome()
driver.get("https://www.google.com")
try:
    element = WebDriverWait(driver, 5).until(
        EC.presence_of_element_located((By.CLASS_NAME, "gLFyf"))
    )
    #요소가 존재하면 텍스트 출력(이 경우에는 입력차이므로 text가 없을 수 있음)
    print('Element found: ', element.get_attribute('title'))
    time.sleep(2)
except Exception as e:
    #예외 발생시 메시지 출력
    print('Element not found', e)
# finally:
#     # driver.close() #브라우저 종료    #드라이버 종료
````

Javascript 선택자 : https://hianna.tistory.com/485 - 비동기식 작동을 위해 사용

1. getElementById()
 - 파라미터로 찾으려는 id를 전달하여, 해당 element를 찾을 수 있습니다. id는 유일한 값이므로, 하나의 element만 리턴합니다.
2. getElementsByClassName()
 - 클래스 이름으로 element를 찾아서,이 클래스 이름을 가지는 모든 element 목록을 리턴합니다.이 함수의 이름을 자세히 보면 getElementsByClassName으로 Element's'가 복수 형태인 것은 이 함수가 목록을 리턴하기 때문입니다.
3. getElementByTagName()
 - 위 코드는 'div' 태그를 가지는 모든 element 목록을 찾아줍니다.이 함수 역시, 다수의 element를 리턴하기 때문에 함수명에 복수형의 'elements'가 포함되어 있습니다.
4. querySelector()
 - DOM에서 원하는 element를 찾기 위해서 querySelector() 를 사용할 수도 있는데, 이 함수는 파라미터로 입력받은 CSS선택자를 사용해서, element를 찾아줍니다.
 - querySelector() 함수는, 파라미터로 입력받은 CSS 선택자로 찾은 여러개의 element 중 첫번째 element를 리턴합니다.
 - 태그 이름으로 element를 찾을 때는 태그명을 문자열로 넘겨줍니다.위 예제는 div태그를 가지는 element 중 첫번째 element를 리턴합니다.
5. querySelectorAll()
 - 사용법은 querySelector() 와 같습니다. 다만, querySelectorAll()은 이름에서 알수 있듯이, CSS선택자(파라미터로 넘겨준)로 찾은 모든 element 목록을 리턴합니다.

````python
from selenium import webdriver

driver = webdriver.Chrome()
driver.get('https://books.toscrape.com/')
##default > div > div > div > div > section > div:nth-child(2) > ol > li:nth-child(1) > article > h3 > a
#('article.product_pod h3 a')
# book_titles = driver.execute_script("return document.querySelector('#default > div > div > div > div > section > div:nth-child(2) > ol > li:nth-child(1) > article > h3 > a').innerText;")
book_titles = driver.execute_script("return document.querySelector('article.product_pod h3 a').innerText;")
print('First book title:', book_titles)
driver.quit()
````

````python
driver = webdriver.Chrome()
driver.get('https://books.toscrape.com/')

#모든 책 제목 가져오기
book_titles = driver.execute_script("""
    var titles = [];
    var bookTitleElements = document.querySelectorAll('article.product_pod h3 a');
    for (var element of bookTitleElements) {
        titles.push(element.innerText);
    }
    return titles;
""")
print('Book titles:', book_titles)
time.sleep(3)
driver.quit()   
````

````python
driver = webdriver.Chrome()
driver.get('https://books.toscrape.com/')

#모든 책 제목 가져오기
book_titles = driver.execute_script("""
    var data = [];
    var elements = document.querySelectorAll('article.product_pod h3 a');
    elements.forEach(function(element) {
        data.push(element.innerText);
    });
    
    return data;
""")
print('Book titles:', book_titles)
time.sleep(3)
driver.quit()   
````
