```python
#전체 적용 코드

import pandas as pd
import numpy as np
import random

random.seed(10)

id = pd.Series(np.arange(1, 1001))
gender = pd.Series(np.random.choice([0,1],1000))
age = pd.Series(np.random.uniform(10,80,1000).astype(int))
region = pd.Series(np.random.uniform(1,10,1000).astype(int))

customer_df = pd.concat([id, gender, age, region],axis=1)
customer_df.columns = ['id', 'gender', 'age', 'region']

#트랜잭션 데이터 프레임

t_no = pd.Series(np.arange(1, 10001))

raw_date = pd.date_range(start='2020-01-01',end='2021-12-31',periods=10000)
dates_only = raw_date.strftime('%Y-%m-%d').tolist()
#date ok
date = pd.Series(dates_only)
# 거래 id ok
id = pd.Series(np.random.uniform(1,1001,10000)).astype(int)
#품목 ok
product = pd.Series(np.random.choice([10,20,30],10000))
#가격 ok
price = pd.Series(np.random.uniform(1000, 50001, 10000))
#수량 ok
quantity = pd.Series(np.random.uniform(1,101, 10000)).astype(int)
#시간 ok
time = pd.Series(np.random.uniform(1,25, 10000)).astype(int)
#amount ok
amount = price * quantity

tran_df = pd.concat([t_no,date,time,id,product,price,quantity,amount],axis=1)
tran_df.columns = ['t_no','date','time','id','product','price','quantity','amount']

total_df = pd.merge( tran_df, customer_df, on='id')

df = total_df.copy()

def adjust_quantity(row):
    if 1 <= row['time'] < 7:
        row['quantity'] *= 0.5
    return row

df = df.apply(adjust_quantity, axis=1)

def age_adjust_quantity(row):
    if  row['age'] >= 60:
        row['quantity'] *= 0.5
    return row

df = df.apply(age_adjust_quantity, axis=1)

def woman_adjust_quantity(row):
    if  row['gender'] == 1:
        row['quantity'] *= 1.2
    return row

df = df.apply(age_adjust_quantity, axis=1)

def chidren_adjust_quantity(row):
    if  row['age'] <= 16:
        row['quantity'] *= 0.3
    return row

df = df.apply(age_adjust_quantity, axis=1)

def top_quantity(row, threshold):
    if row['amount'] > threshold:
        row['quantity'] *= 2
    return row

threshold = df['amount'].quantile(0.90)

df = df.apply(lambda row: top_quantity(row, threshold), axis=1)

df
```
