### 매년 구매고객에 대한 데이터 프레임으로 전환
- 2020년도를 기준으로 하여 고객, 나이 ,지역, 매출총합의 데이터 프레임 생성
- 2021년도를 기준으로 하여 고객, 나이 ,지역, 매출총합의 데이터 프레임 생성
- 양 데이터 프레임을 통합하여 분석용 데이터셋 구성


```python
df['year'] = pd.to_datetime(df['date']).dt.year

df_2020 = df[df['year'] == 2020]
df_2021 = df[df['year'] == 2021]

df_2020_sum = df_2020.groupby(['cusno', 'age', 'region', 'product'])['amount'].sum().unstack(fill_value=0).reset_index()
df_2021_sum = df_2021.groupby(['cusno', 'age', 'region', 'product'])['amount'].sum().unstack(fill_value=0).reset_index()

df_2020_sum.columns = ['cusno', 'age', 'region', '20product10', '20product20', '20product30']
df_2021_sum.columns = ['cusno', 'age', 'region', '21product10', '21product20', '21product30']

customer_df = pd.merge(df_2020_sum, df_2021_sum, on=['cusno', 'age', 'region'], how='outer').fillna(0)

customer_df
```

```python
#물건 종류별 매출 변동률
#unstack() 을 이용해서 DataFrame의 모양을 다루기 쉽게 바꿀 수 있다.
#groupby 한 결과를 group 별로 더 조작해야 될 때 유용하다.
change = df.groupby(['year', 'product'])['amount'].sum().unstack()
change = change.pct_change().fillna(0) * 100
change
```
