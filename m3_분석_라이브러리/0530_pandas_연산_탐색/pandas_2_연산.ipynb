```python
import pandas as pd

st1 = pd.Series({'국어':100 ,'영어':80 , '수학':90})
st2 = pd.Series({'국어':90 ,'영어':80 , '수학':80})

print(st1, '\n')
print(st2)
```

```python
pd.DataFrame(st1+st2)
```

```python
#두 학생의 과목별 점수로 사칙연산 수행
add = st1 + st2
sub = st1 - st2
mul = st1 * st2
div = round((st1 / st2),2)

df = pd.concat([add,sub,mul,div],axis=1)
print(df)
```

데이터 집계(agg),
- 집계 연산을 처리하는 사용자 정의 함수를 그룹 객체에 적용하려면 agg() 메서드 사용
- 모든열에 여러 함수를 매핑 : group객체.agg([함수1,함수2,함수3,…])
- 각 열마다 다른 함수를 매핑 : group객체.agg({‘열1’: 함수1, ‘열2’:함수2, …})

변환(transform)
- 원소의 본래 행 인덱스와 열 이름을 기준으로 연산 결과를 반환
- 데이터 변환 연산 : group객체.transform(매핑함수)

필터링(filter)
- 그룹 객체에 filter() 메서드를 적용할 때 조건식을 가진 함수를 전달하면 조건이 참인 그룹만을 남김
- 그룹 객체 필터링 : group 객체.filter(조건 함수식)

객체에 함수 매핑(apply)
 - 판다스 객체의 개별 원소를 특정 함수에 일대일로 매핑. 사용자가 원하는 대부분의 연산을 그룹 객체에 적용
 - group 객체.apply(매핑함수)

```python
import pandas as pd
import seaborn as sns

# titanic 데이터셋에서 age, sex 등 5개 열을 선택하여 데이터프레임 만들기
titanic = sns.load_dataset('titanic')
df = titanic.loc[:, ['age','class', 'fare', 'survived']]

# class 열을 기준으로 분할
grouped = df.groupby(['class'])
```

```python
mean_all = grouped.mean()
print(mean_all)
print()
print(type(mean_all))
```

```python
mean_all.index.name = None
print(mean_all)
```

```python
#각 그룹에 대한 모든 열의 표준편차를 집계하여 데이터프레임으로 반환
#표준편차- 평균에서 얼마나 벗어나 있나
#1등석 요금 차이가 심함
std_all = grouped.std()
print(std_all)
print()
print(type(std_all))
```

```python
#각 그룹에 대한 fare열의 표준편차를 집계하여 시리즈로 반환 한개의 열만 뽑으니 시리즈로 나옴
std_fare = grouped.fare.std()
print(std_fare)
print()
print(type(std_fare))
```

```python
#각 그룹에 대한 age열의 표준편차를 집계하여 시리즈로 반환
std_age = grouped.age.std()
print(std_age)
print()
print(type(std_age))
```

```python
#그룹 객체에 agg()메소드 적용 - 사용자 정의 함수를 인수로 전달
def min_max(x):     #최대값 - 최소값
    return x.max() - x.min()

agg_minmax = grouped.agg(min_max)
print(agg_minmax.head())
```

```python
# 모든열에 여러 함수를 매핑 : group객체.agg([함수1,함수2,함수3,…])
# 각 열마다 다른 함수를 매핑 : group객체.agg({‘열1’: 함수1, ‘열2’:함수2, …})

agg_all = grouped.agg(['min','max', 'mean'])
print(agg_all.head())
```

```python
agg_all = grouped.agg(['min','max', min_max, 'mean'])
print(agg_all.head())
```

```python
#각 열마다다른 함수를 적용하여 집계
agg_sep = grouped.agg({'fare':['min','max'],'age':'mean'})
print(agg_sep.head())
```

#### 함수 매핑
커스텀 함수(custom function)를 DataFrame에 적용하려면 map함수, apply함수, applymap함수를 사용
- map 함수는 Series의 각 요소에 함수를 적용.
- apply 함수는 Series 또는 DataFrame의 행이나 열에 함수를 적용.
- applymap 함수는 DataFrame의 각 요소에 함수를 적용.

```python
import pandas as pd
#시리즈 생성
s = pd.Series([1,2,3,4,5])
#각 요소에 2 더하는 함수
s_mapped = s.map(lambda x: x + 2)
print(s_mapped)
```

```python
#series에 적용: Series생성
```

```python
#
df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6],
    'C': [7, 8, 9]
})
print(df)
print()
#각 열에 대해 최대값을 구하는 함수 , 열을 판단하려면 행으로 봐야함 axis=0
df_applied_col = df.apply(lambda x: x.max(),axis=0)
print(df_applied_col)
#각 행에 대해 합계를 구하는 함수, 행을 판단하기 위해 열로 봐야함 axis=1
df_applied_row = df.apply(lambda x: x.sum(), axis=1)
print(df_applied_row)
```

```python
#applymap함수는 데이터프레임에만 적용되며 각 요서에 함수를 적용
df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6],
    'C': [7, 8, 9]
})

df_applymapped = df.applymap(lambda x: x + 2)
print(df_applymapped)
```

```python
#표준화: 데이터의 평균을 0, 표준편차를 1로 변환하는 과정. 이과정을 통해
#데이터의 스케일을 조정하고 비교 가능한 형태로 만듬

def z_score(x):
    return (x - x.mean()) / x.std()
```

```python
df_zscore = df.apply(lambda x:z_score(x))
df_zscore
```

####Pivot
pivot 함수의 구성 형식
- index: 새로운 DataFrame에서 인덱스로 사용할 기존 열 또는 열들의 이름.
- columns: 새로운 DataFrame에서 열로 사용할 기존 열 또는 열들의 이름.
- values: 피벗할 때 사용할 값이 있는 열의 이름.

기능
- 긴 형식의 데이터를 넓은 형식으로 변환하여 특정 차원에서 데이터를 재구성.
(컬럼과 컬럼을 비교)
- 피벗된 데이터를 통해 시간 경과에 따른 변화나 여러 카테고리의 비교를 쉽게 할 수 있다.
- 넓은 형식의 데이터는 시각화 도구에서 더 쉽게 다룰 수 있다.
-알고리즘을 위한 데이터셋을 만들기 위해 피봇을 쓰는건 아님



```python
#날짜를 인덱스로 도시를 열로, 온도를 값으로 사용하여 피벗테이블 생성
import pandas as pd

data = {
    'date': ['2021-01-01', '2021-01-01', '2021-01-02', '2021-01-02'],
    'city': ['New York', 'Los Angeles', 'New York', 'Los Angeles'],
    'temperature': [32, 75, 30, 78]
}
df = pd.DataFrame(data)
print(df)
print()
#피벗 테이블 생성  pivot()
pivot_df = df.pivot(index='date', columns='city', values='temperature')
print(pivot_df)
```

```python
#중복된 값을 평균으로 처리하여 피벗 테이블을 생성
import pandas as pd

data = {
    'date': ['2021-01-01', '2021-01-01', '2021-01-01','2021-01-02', '2021-01-02'],
    'city': ['New York', 'Los Angeles', 'New York','New York','Los Angeles'],
    'temperature': [32, 75, 30,30, 78]
}
df = pd.DataFrame(data)
#피벗 테이블 생성(중복값의 평균 계산)   pivot_table()
pivot_df = df.pivot_table(index='date', columns='city', values='temperature', aggfunc='mean')
print(pivot_df)
```

```python
#다중 인덱스와 다중 값 사용: 날짜와 도시를 기준으로 온도와 습도를 모두 포함한 피벗 테이블을 생성

data = {
    'date': ['2021-01-01', '2021-01-01', '2021-01-02', '2021-01-02'],
    'city': ['New York', 'Los Angeles', 'New York', 'Los Angeles'],
    'temperature': [32, 75, 30, 78],
    'humidity': [80, 20, 85, 15]
}

df = pd.DataFrame(data)
print(df, '\n')

pivot_multi_df = df.pivot(index='date', columns='city')
print(pivot_multi_df)
```

Q. 주어진 매출 데이터에서, 각 연도와 각 제품별로 총 매출액을 계산하세요.

```python
import pandas as pd

data = {
    'Year': [2020, 2020, 2021, 2021, 2022, 2022, 2022, 2023, 2023],
    'Product': ['A', 'B', 'A', 'B', 'A', 'B', 'C', 'A', 'C'],
    'Sales': [100, 150, 200, 250, 300, 350, 400, 500, 450]
}
df = pd.DataFrame(data)
print(df, '\n')
pivot_df = df.pivot(index='Year', columns='Product', values='Sales')
pivot_df.loc['Sum'] = pivot_df.apply('sum')
print(pivot_df)
```

```python
import pandas as pd

data = {
    'Year': [2020, 2020, 2021, 2021, 2022, 2022, 2022, 2023, 2023],
    'Product': ['A', 'B', 'A', 'B', 'A', 'B', 'C', 'A', 'C'],
    'Sales': [100, 150, 200, 250, 300, 350, 400, 500, 450]
}
df = pd.DataFrame(data)
print(df, '\n')
pivot_df = df.pivot_table(index='Year', columns='Product', values='Sales', aggfunc='sum', fill_value=0)
print(pivot_df)
```
